# Visual Review: The 'Big Three' Distributions

This notebook provides a quick visual reference for the PMF graphs of the Binomial, Poisson, and Negative Binomial distributions. The notes highlight the key features to remember for the exam.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, nbinom

# Set a consistent style for the plots
plt.style.use('seaborn-v0_8-whitegrid')
```

## 1. Binomial Distribution
*The Story: Number of successes in a **fixed `n` trials**.*

```python
n, p = 20, 0.7  # 20 trials, 70% chance of success
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title(f'Binomial PMF (n={n}, p={p})', fontsize=16)
plt.xlabel('Number of Successes (k)', fontsize=12)
plt.ylabel('P(X = k)', fontsize=12)
plt.xticks(x)
plt.show()
```

### Key Features to Note (Binomial):

1.  **Finite X-Axis:** The graph has a hard stop. You can't have more than `n` successes (here, 20). The number of possible outcomes is limited.
2.  **Symmetry/Skew:** The graph's shape is determined by `p`.
    *   If `p = 0.5`, the graph is perfectly symmetric.
    *   If `p > 0.5` (like our `p=0.7`), the probability mass piles up on the right, making it **skewed left**.
    *   If `p < 0.5`, the mass piles up on the left, making it **skewed right**.
3.  **The Peak:** The most likely outcome (the peak) is always around `n * p`. Here, `20 * 0.7 = 14`, which is exactly where the graph peaks.

## 2. Poisson Distribution
*The Story: Number of events in a **fixed interval** of time/space.*

```python
lambda_val = 8  # Average of 8 events per interval
x = np.arange(0, 20) # We only plot up to 20 for visibility
pmf = poisson.pmf(x, lambda_val)

plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='salmon', edgecolor='black')
plt.title(f'Poisson PMF (={lambda_val})', fontsize=16)
plt.xlabel('Number of Events (k)', fontsize=12)
plt.ylabel('P(X = k)', fontsize=12)
plt.xticks(x)
plt.show()
```

### Key Features to Note (Poisson):

1.  **Infinite X-Axis (Theoretically):** The graph has a long tail to the right. While the probabilities get tiny, there is no theoretical upper limit to the number of events that could occur. This is a key difference from the Binomial.
2.  **Always Skewed Right:** The Poisson distribution is always skewed right, but it becomes more symmetric and bell-shaped as `` gets larger.
3.  **The Peak:** The most likely outcome is always at or right next to the average rate, ``. Here, the peak is at `k=8`.

## 3. Negative Binomial Distribution
*The Story: Number of **failures** before reaching a **fixed `r` successes**.*

```python
r, p = 5, 0.2  # Waiting for 5 successes, with a 20% chance of success
x = np.arange(0, 41) # Plotting the number of failures
pmf = nbinom.pmf(x, r, p)

plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='lightgreen', edgecolor='black')
plt.title(f'Negative Binomial PMF (r={r}, p={p})', fontsize=16)
plt.xlabel('Number of Failures (k)', fontsize=12)
plt.ylabel('P(X = k)', fontsize=12)
plt.show()
```

### Key Features to Note (Negative Binomial):

1.  **Infinite X-Axis (Theoretically):** Like the Poisson, this graph has a long right tail. It's theoretically possible to have an incredibly long streak of failures before you hit your target number of successes.
2.  **Always Skewed Right:** This distribution is characterized by its right skew. Low numbers of failures are most likely, but there's always a chance of a very long, unlucky run.
3.  **Starts at k=0:** The x-axis represents failures. It's possible to have zero failures (i.e., you get all `r` successes in a row), so the graph always starts at `k=0`.
