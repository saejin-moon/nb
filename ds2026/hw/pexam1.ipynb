**2.12 Create 9 plots of the binomial distribution probability mass function with N=10 and p = .1, .2, …, .9**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

N = 10
p_values = np.arange(0.1, 1.0, 0.1)
x = np.arange(0, N + 1)

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel() # Flatten the 2D array of axes

for i, p in enumerate(p_values):
    pmf = binom.pmf(x, N, p)
    axes[i].bar(x, pmf, label=f'p = {p:.1f}')
    axes[i].set_title(f'Binomial PMF (N=10, p={p:.1f})')
    axes[i].set_xlabel('Number of Successes')
    axes[i].set_ylabel('Probability')
    axes[i].set_xticks(x)

plt.tight_layout()
plt.show()
````

**2.14 Suppose electric bikes represent 0.1 of all bikes among UVA students. If a researcher were to stand at the Emmett and Ivy intersection to count bikes, what is the probability that 30 or fewer regular bikes will be observed before the 5 e-bike is observed.**

This is a negative binomial problem.
*   p = 0.1.
*   k = 5.
*   A "failure" is observing a regular bike. We want the probability that the number of failures is 30 or fewer.

We need to calculate the cumulative probability P(X ≤ 30), where X is the number of failures.

````python
from scipy.stats import nbinom

k = 5
p = 0.1
x = 30

prob_cdf = nbinom.cdf(x, k, p)
print(f"{prob_cdf:.4f}")
````

**2.16 Suppose accidents at the Emmett Ivy intersection occur at a rate of 5 per year. If accidents follow the Poisson distribution, what is probability of observing 3 accidents in 9 months.**

First, we adjust the rate (λ) for the new time interval.
*   λ = 5
*   9 months = 0.75 years
*   λ' = 5 * 0.75 = 3.75 accidents per 9 months.

Now we want to find P(X=3) for a Poisson distribution with λ' = 3.75.

````python
from scipy.stats import poisson

lambda_adj = 3.75
k = 3

prob = poisson.pmf(k, lambda_adj)
print(f"{prob:.4f}")
````

**2.17 Consider a class of 15 students, of which 9 are right handed, 4 are left handed, and 2 are ambidextrous. What is the probability that a team of 8 students has no left handed students if teams are created randomly.**

This is a hypergeometric probability problem.
*   Population size, N = 15.
*   Number of items of interest (left-handed), K = 4.
*   Number of non-interest items (right-handed or ambidextrous), N-K = 11.
*   Sample size, n = 8.
*   Number of items of interest in sample (left-handed), k = 0.

The formula is $P(X=k) = [C(K, k) * C(N-K, n-k)] / C(N, n)$.
$P(X=0) = [C(4, 0) * C(11, 8)] / C(15, 8)$

````python
from scipy.special import comb

# C(4, 0) = 1
# C(11, 8) = 165
# C(15, 8) = 6435

prob = (comb(4, 0) * comb(11, 8)) / comb(15, 8)
print(f"{prob:.4f}")
````

**3.1 Write a simulation that estimates the probability from question 2.17. Use only 200 replicates.**

````python
import numpy as np

population = ['R'] * 9 + ['L'] * 4 + ['A'] * 2

n_replicates = 200
sample_size = 8
success_count = 0

for _ in range(n_replicates):
    team = np.random.choice(population, size=sample_size, replace=False)
    if 'L' not in team:
        success_count += 1

estimated_probability = success_count / n_replicates

print(f"Theoretical probability: {165/6435:.4f}")
print(f"Simulated probability from {n_replicates} replicates: {estimated_probability:.4f}")
````

**3.2 Calculate the simulation error from problem 3.1.**

The standard error of a simulated proportion is calculated as `sqrt(p * (1 - p) / n)`, where `p` is the estimated probability and `n` is the number of replicates.

Using the example output from 3.1 where `p_est = 0.0256` and `n = 200`:

````python
p_est = 0.0256
n = 200
simulation_error = np.sqrt(p_est * (1 - p_est) / n)

print(f"Simulation error: {simulation_error:.4f}")
````
