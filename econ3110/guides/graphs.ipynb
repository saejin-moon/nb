```python
import matplotlib.pyplot as plt
import numpy as np

# Setup Figure
fig, axes = plt.subplots(2, 3, figsize=(24, 14))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# =======================================================
# 1. CONSUMER OPTIMIZATION (The Wallet & The Stomach)
# =======================================================
ax = axes[0, 0]
x = np.linspace(0.1, 20, 100)
M, px, py = 100, 10, 10
# Budget Line
budget = (M - px*x) / py
# Indifference Curve (U = xy = 25)
u_curve = 25 / x

ax.plot(x, budget, 'r-', linewidth=2, label='Budget Constraint')
ax.plot(x, u_curve, 'b-', linewidth=2, label='Indifference Curve (IC)')
ax.plot(5, 5, 'ko', markersize=10) # Optimum

# Annotations
ax.text(1, 2, f'Slope = -$P_x/P_y$\n(Opportunity Cost)', color='red', fontsize=10)
ax.text(8, 8, f'Slope = MRS\n(Willingness to Pay)', color='blue', fontsize=10)
ax.annotate('Optimal Bundle\n$MRS = P_x/P_y$', xy=(5, 5), xytext=(8, 12),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_title('1. Consumer Choice', fontsize=14, fontweight='bold')
ax.set_xlabel('Quantity of Good X', fontsize=12)
ax.set_ylabel('Quantity of Good Y', fontsize=12)
ax.set_xlim(0, 15); ax.set_ylim(0, 15)
ax.legend(loc='upper right')

# =======================================================
# 2. PRODUCER COST MINIMIZATION (The Kitchen)
# =======================================================
ax = axes[0, 1]
L = np.linspace(0.1, 20, 100)
# Isocost (Total Cost = 100, w=10, r=10)
isocost = (100 - 10*L) / 10
# Isoquant (Q = sqrt(LK) = 5 -> K = 25/L)
isoquant = 25 / L

ax.plot(L, isocost, 'r-', linewidth=2, label='Isocost Line')
ax.plot(L, isoquant, 'g-', linewidth=2, label='Isoquant (Tech)')
ax.plot(5, 5, 'ko', markersize=10)

# Annotations
ax.text(1, 2, f'Slope = -$w/r$\n(Input Price Ratio)', color='red', fontsize=10)
ax.text(8, 8, f'Slope = MRTS\n(MP_L / MP_K)', color='green', fontsize=10)
ax.annotate('Cost Min Bundle\n$MRTS = w/r$', xy=(5, 5), xytext=(8, 12),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_title('2. Producer Cost Min', fontsize=14, fontweight='bold')
ax.set_xlabel('Labor (L)', fontsize=12)
ax.set_ylabel('Capital (K)', fontsize=12)
ax.set_xlim(0, 15); ax.set_ylim(0, 15)
ax.legend(loc='upper right')

# =======================================================
# 3. MONOPOLY (The Price Maker)
# =======================================================
ax = axes[0, 2]
Q = np.linspace(0, 12, 100)
Demand = 10 - Q
MR = 10 - 2*Q
MC = np.full_like(Q, 2)

ax.plot(Q, Demand, 'b-', linewidth=2, label='Demand (AR)')
ax.plot(Q, MR, 'g--', linewidth=2, label='Marginal Revenue (MR)')
ax.plot(Q, MC, 'r-', linewidth=2, label='Marginal Cost (MC)')

# Equilibrium
ax.vlines(4, 0, 6, linestyle=':') # Qm to Price
ax.hlines(6, 0, 4, linestyle=':') # Price Axis
ax.plot(4, 2, 'ko') # MR=MC intersection
ax.plot(4, 6, 'bo') # Price point

# Annotations
ax.text(5, 8, 'MR has 2x Slope\nof Demand', color='green', fontsize=9)
ax.annotate('Profit Max Q\n(MR = MC)', xy=(4, 2), xytext=(6, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Monopoly Price\n(From Demand Curve!)', xy=(4, 6), xytext=(6, 8),
            arrowprops=dict(facecolor='blue', shrink=0.05))
# DWL
ax.fill_between(Q, Demand, 2, where=(Q>=4)&(Q<=8), color='gray', alpha=0.3)
ax.text(4.5, 3, 'Deadweight\nLoss', fontsize=10)

ax.set_title('3. Monopoly Equilibrium', fontsize=14, fontweight='bold')
ax.set_xlabel('Quantity (Q)', fontsize=12)
ax.set_ylabel('Price ($)', fontsize=12)
ax.set_xlim(0, 10); ax.set_ylim(0, 11)
ax.legend(loc='upper right')

# =======================================================
# 4. COURNOT DUOPOLY (The Reaction War)
# =======================================================
ax = axes[1, 0]
q2 = np.linspace(0, 100, 100)
# Reaction Functions (Derived from P = 100-Q, MC=10)
# q1 = 45 - 0.5*q2
rf1 = 45 - 0.5 * q2
# q2 = 45 - 0.5*q1 -> q1 = 90 - 2*q2 (for plotting)
rf2_plot = 90 - 2 * q2

ax.plot(rf1, q2, 'b-', linewidth=2, label="Firm 1's Best Response")
ax.plot(q2, 45 - 0.5*q2, 'r-', linewidth=2, label="Firm 2's Best Response")
ax.plot(30, 30, 'ko', markersize=10)

# Annotations
ax.text(5, 35, 'If Firm 2 produces 0,\nFirm 1 is Monopoly (45)', color='blue', fontsize=9)
ax.annotate('Nash Equilibrium\n(Intersection)', xy=(30, 30), xytext=(50, 50),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_title('4. Cournot Oligopoly', fontsize=14, fontweight='bold')
ax.set_xlabel('Quantity Firm 1 ($q_1$)', fontsize=12)
ax.set_ylabel('Quantity Firm 2 ($q_2$)', fontsize=12)
ax.set_xlim(0, 60); ax.set_ylim(0, 60)
ax.legend(loc='upper right')

# =======================================================
# 5. GENERAL EQUILIBRIUM (Edgeworth Box)
# =======================================================
ax = axes[1, 1]
xA = np.linspace(0.1, 9.9, 100)
# Contract Curve (yA = xA for Cobb Douglas)
ax.plot(xA, xA, 'k-', linewidth=3, label='Contract Curve')
# Indifference Curves
ax.plot(xA, 25/xA, 'b-', alpha=0.5, label="Ash's IC")
y_misty = 10 - 25/(10-xA) # Inverted IC for Misty
ax.plot(xA, y_misty, 'r-', alpha=0.5, label="Misty's IC")
ax.plot(5, 5, 'go', markersize=10)

# Annotations
ax.text(1, 8, 'Ash Origin (0,0)', color='blue')
ax.text(7, 1, "Misty Origin (Max, Max)", color='red')
ax.annotate('Pareto Efficient\n($MRS_A = MRS_B$)', xy=(5, 5), xytext=(1, 5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_title('5. Edgeworth Box', fontsize=14, fontweight='bold')
ax.set_xlabel("Ash's Good X", fontsize=12)
ax.set_ylabel("Ash's Good Y", fontsize=12)
ax.set_xlim(0, 10); ax.set_ylim(0, 10)
ax.legend(loc='upper right')

# =======================================================
# 6. EXTERNALITY (Pollution)
# =======================================================
ax = axes[1, 2]
Q = np.linspace(0, 12, 100)
Demand = 12 - Q        # Social Benefit
PMC = 2 + Q            # Private Supply
SMC = 4 + Q            # Social Cost (PMC + 2)

ax.plot(Q, Demand, 'b-', linewidth=2, label='Demand (Benefit)')
ax.plot(Q, PMC, 'g-', linewidth=2, label='Private Cost (Supply)')
ax.plot(Q, SMC, 'r-', linewidth=2, label='Social Cost (Supply + Tax)')

# Equilibria
Q_mkt = 5; P_mkt = 7
Q_soc = 4; P_soc = 8
ax.plot(Q_mkt, P_mkt, 'ko')
ax.plot(Q_soc, P_soc, 'go')

# Annotations
ax.vlines(Q_soc, 6, 8, color='orange', linewidth=3)
ax.text(Q_soc + 0.5, 7, 'Pigouvian Tax\n(Vertical Distance)', color='orange', fontweight='bold')
ax.annotate('Social Optimum\n(SMC = Demand)', xy=(Q_soc, P_soc), xytext=(8, 10),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Private Market\n(PMC = Demand)', xy=(Q_mkt, P_mkt), xytext=(8, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_title('6. Negative Externality', fontsize=14, fontweight='bold')
ax.set_xlabel('Quantity', fontsize=12)
ax.set_ylabel('Price', fontsize=12)
ax.legend(loc='upper right')

plt.show()
```
