### **Chapter 1: The Budget Constraint**

#### **(1) Intuition First**
Welcome to **The Eco-Nom-Nom Food Truck**. Before we can analyze what our customers *want* to eat, we must understand what they can *afford*.

Imagine a student customer, Ash. He has a fixed amount of money in his pocket (`$`30). He wants to buy Burgers ($B$) and Fries ($F$).
*   Every time he buys a Burger, he has less money for Fries.
*   The "trade-off" is determined by the price stickers on the menu.
*   If he spends *all* his money, he is on the "edge" of his financial possibilities.

In economics, we don't just look at money. We also look at **Time Constraints**. Ash has only 24 hours a day. If he spends time eating, he has less time for studying. The logic is identical: resources are scarce.

#### **(2) Formal Definition (Translated)**
The **Budget Set** represents every possible combination of goods a consumer can afford.
The **Budget Line** represents the combinations where the consumer spends exactly all their income.

If there are two goods, $x_1$ and $x_2$, with prices $p_1$ and $p_2$, and the consumer has income $M$:
$$p_1 x_1 + p_2 x_2 \leq M$$

*   **Slope of the line:** $-\frac{p_1}{p_2}$. This is the **Relative Price**. It tells you how many units of Good 2 you *must* give up to buy one more unit of Good 1.

#### **(3) Graphs & Relationships**

The following code generates a standard Budget Line graph.

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
M = 30      # Income
p_B = 10    # Price of Burgers
p_F = 5     # Price of Fries

# Calculate intercepts
max_B = M / p_B  # Max burgers if 0 fries
max_F = M / p_F  # Max fries if 0 burgers

# Create points for the line
burgers = np.linspace(0, max_B, 100)
fries = (M - p_B * burgers) / p_F

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(burgers, fries, label=f'Budget Line: ${p_B}B + ${p_F}F = ${M}', color='blue', linewidth=2)
plt.fill_between(burgers, fries, 0, alpha=0.1, color='blue', label='Budget Set (Affordable)')

# Formatting
plt.title(f'Ash\'s Budget Constraint (Income = ${M})')
plt.xlabel('Quantity of Burgers (B)')
plt.ylabel('Quantity of Fries (F)')
plt.xlim(0, max_B + 1)
plt.ylim(0, max_F + 1)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

**Key Relationships from the Graph:**
*   **Vertical Intercept:** $M/p_F$ (All money spent on Fries).
*   **Horizontal Intercept:** $M/p_B$ (All money spent on Burgers).
*   **Slope:** Negative. Calculated as $-\frac{Price_{Burgers}}{Price_{Fries}}$.
    *   *Real World Translation:* The "Opportunity Cost" of a Burger is 2 Fries (since `$`10 / `$`5 = 2).

#### **(4) Key Takeaways**
*   **Scarcity:** You can't have everything. You are constrained by $M$.
*   **Trade-offs:** The slope tells you the market trade-off. To get more $x$, you must give up $y$.
*   **Shifts:**
    *   If **Income ($M$)** changes: The line shifts parallel (out for richer, in for poorer).
    *   If **Price ($p$)** changes: The line **pivots** or rotates. If Burgers get expensive, the max Burgers you can buy drops, but max Fries stays the same.

#### **(5) Worked Example**
**Scenario:** Ash has $\$30$.
*   Price of Burgers ($P_B$) = $\$10$.
*   Price of Fries ($P_F$) = $\$5$.

**Step 1: Find the extremes.**
*   All Burgers: $30 / 10 = 3$ Burgers.
*   All Fries: $30 / 5 = 6$ Fries.

**Step 2: Find the Slope.**
*   Slope = $-P_B / P_F = -10 / 5 = -2$.
*   *Interpretation:* To buy 1 Burger, Ash must give up 2 Fries.

**Step 3: Equation.**
$$10B + 5F = 30$$

**Step 4: A Change.**
The truck owner feels generous. He lowers the price of Burgers to $\$5$.
*   New Slope: $-5 / 5 = -1$.
*   Ash can now afford 6 Burgers if he buys only Burgers. The line pivots out.

### **Chapter 2: Consumer Preferences**

#### **(1) Intuition First**
Back to **The Eco-Nom-Nom Food Truck**. In Chapter 1, we looked at Ash's wallet (constraints). Now, we look at his stomach (desires).

Ash stands in front of the menu. He sees **Burgers ($B$)** and **Fries ($F$)**.
*   **Ranking:** Ash can look at two different trays of foodsay, Tray A (2 Burgers, 1 Fry) and Tray B (1 Burger, 5 Fries)and tell you which he prefers, or if he is indifferent.
*   **More is Better:** Ash is hungry. If Tray C has *more* Burgers and *more* Fries than Tray A, he definitely wants Tray C. This is called **Monotonicity**.
*   **Variety is Good:** Ash usually prefers a mix. He would rather have a meal with some Burgers and some Fries than a tray with 10 Burgers and 0 Fries. This preference for "averages over extremes" is called **Convexity**.

We map these desires using **Indifference Curves**. Imagine a topographic map of a mountain. Every point on a single contour line is at the same height. Similarly, every bundle of food on a single **Indifference Curve** gives Ash the exact same level of happiness (Utility).

#### **(2) Formal Definition (Translated)**

*   **Utility Function ($U(x_1, x_2)$):** A mathematical way to assign a number to happiness.
    *   *Crucial Note:* The number itself doesn't matter (Ordinality). $U=100$ isn't "twice as happy" as $U=50$. It just means the first bundle is *better* than the second.
    *   You can square a utility function, multiply it by 10, or add 5, and it represents the *exact same preferences* as long as the ranking order stays the same.

*   **Marginal Rate of Substitution (MRS):** The slope of the Indifference Curve.
    *   Formula: $MRS = \frac{MU_1}{MU_2}$ (where $MU$ is Marginal Utility).
    *   *Translation:* How many Fries is Ash willing to give up to get **one more** Burger and stay equally happy?
    *   If Ash has 100 Fries and 1 Burger, he'd trade a *lot* of Fries for a second Burger (High MRS).
    *   If Ash has 10 Burgers and 1 Fry, he wouldn't give up his last Fry for another Burger (Low MRS). This is **Diminishing MRS**.

#### **(3) Graphs & Relationships**

There are three main "personalities" of customers we see at the truck. We will visualize them below.

```python
import matplotlib.pyplot as plt
import numpy as np

# Grid setup
x = np.linspace(0.1, 10, 100)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Cobb-Douglas (Standard mix)
# U = x*y -> y = U/x
U_levels = [5, 10, 20]
for u in U_levels:
    axes[0].plot(x, u/x, label=f'U={u}', linewidth=2)
axes[0].set_title("Standard Tastes (Cobb-Douglas)\nVariety is good")
axes[0].set_xlabel("Burgers")
axes[0].set_ylabel("Fries")
axes[0].set_xlim(0, 10)
axes[0].set_ylim(0, 10)
axes[0].grid(True, linestyle='--', alpha=0.3)

# 2. Perfect Substitutes
# U = x + y -> y = U - x
for u in U_levels:
    axes[1].plot(x, u - x, label=f'U={u}', linewidth=2)
axes[1].set_title("Perfect Substitutes\nI don't care, just give me caffeine")
axes[1].set_xlabel("Coke")
axes[1].set_ylabel("Pepsi")
axes[1].set_xlim(0, 10)
axes[1].set_ylim(0, 10)
axes[1].grid(True, linestyle='--', alpha=0.3)

# 3. Perfect Complements
# U = min(x, y) -> L-shaped
for u in [2, 4, 6]:
    axes[2].vlines(u, u, 10, colors='blue', linewidth=2) # Vertical part
    axes[2].hlines(u, u, 10, colors='blue', linewidth=2, label=f'U={u}') # Horizontal part
axes[2].set_title("Perfect Complements\nMust go together")
axes[2].set_xlabel("Left Shoes")
axes[2].set_ylabel("Right Shoes")
axes[2].set_xlim(0, 10)
axes[2].set_ylim(0, 10)
axes[2].grid(True, linestyle='--', alpha=0.3)

plt.show()
```

**Graph Breakdown:**
1.  **Cobb-Douglas (Curved):** Ash likes a balance. As he gets more Burgers, he becomes less willing to give up Fries. The curve "bows" inward (Convex).
2.  **Perfect Substitutes (Straight Line):** Ash is buying soda. He doesn't care if it's Coke or Pepsi. He is willing to trade 1 Coke for 1 Pepsi always. The slope (MRS) is constant.
3.  **Perfect Complements (L-Shaped):** Ash is buying Hot Dogs and Buns. Having 50 Hot Dogs and 1 Bun is useless. He only cares about the number of *pairs*.

#### **(4) Key Takeaways**
*   **Indifference Curves (ICs)** represent "mountains" of happiness. You want to be on the highest possible curve.
*   **ICs never cross.** (If they did, logic would break).
*   **MRS** is the slope of the IC. It tells you the consumer's *internal* trade-off rate (willingness to pay).
*   **Utility Functions** are just tools to rank choices. $U = x \cdot y$ implies the same behavior as $U = 100 \cdot (x \cdot y)^2$.

#### **(5) Worked Example**
**Scenario:** Ash has utility function $U(B, F) = B \cdot F$.
He currently has Bundle A: (2 Burgers, 8 Fries).
*   Current Utility: $2 \times 8 = 16$.

**Question:** Would he trade Bundle A for Bundle B: (4 Burgers, 4 Fries)?
*   Utility of B: $4 \times 4 = 16$.
*   **Result:** Yes, he is **Indifferent**. These two points lie on the exact same Indifference Curve.

**Question:** What about Bundle C: (3 Burgers, 6 Fries)?
*   Utility of C: $3 \times 6 = 18$.
*   **Result:** $18 > 16$. He strictly prefers C. C is on a "higher" indifference curve (further to the top-right).

**Calculating MRS at Bundle A (2, 8):**
*   How many Fries would he give up for 1 more Burger?
*   Rough approximation: To keep $U=16$, if $B$ becomes 3, $F$ must be $16/3 = 5.33$.
*   He gives up $8 - 5.33 = 2.66$ Fries.
*   *Calculus shortcut:* For $U = x \cdot y$, $MRS = y/x$.
*   At $(2, 8)$, $MRS = 8/2 = 4$. He values 1 Burger as much as 4 Fries right now.

### **Chapter 3: The Optimal Choice**

#### **(1) Intuition First**
Back at **The Eco-Nom-Nom Food Truck**, Ash has his wallet (Budget Constraint, Ch 1) and his appetite (Preferences, Ch 2). Now he has to order.

He wants to reach the "highest" mountain of happiness he can, but he is fenced in by his budget line.
*   **The Sweet Spot:** Imagine Ash is walking along his budget line (the fence).
    *   If he is at a point where he has mostly Fries, he is willing to trade a *lot* of Fries for one Burger (High MRS). But the truck only charges `$`10 for a Burger vs `$`5 for Fries (Price Ratio = 2).
    *   **The Logic:** "I value a Burger at 4 Fries (MRS), but the truck only charges me 2 Fries (Price Ratio). It's a bargain! I should buy more Burgers."
    *   **The Stop Sign:** He keeps trading Fries for Burgers until his personal valuation (MRS) exactly matches the market price (Price Ratio). At this point, trading any further would make him worse off. This is the point of **Tangency**.

#### **(2) Formal Definition (Translated)**
The **Optimal Bundle** $(x_1^*, x_2^*)$ is the affordable combination that maximizes Utility.

For standard preferences (smooth curves), this happens where the Indifference Curve touches the Budget Line at exactly one point (Tangency).

**The Golden Rule of Optimization:**
$$MRS_{1,2} = \frac{p_1}{p_2}$$

Or, rewritten as the **"Bang-Per-Buck" Principle**:
$$\frac{MU_1}{p_1} = \frac{MU_2}{p_2}$$
*   *Translation:* The happiness you get from the last dollar spent on Good 1 must equal the happiness from the last dollar spent on Good 2. If $MU_1/p_1 > MU_2/p_2$, you should take money away from Good 2 and spend it on Good 1.

#### **(3) Graphs & Relationships**

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
M = 100      # Income
px = 10      # Price X
py = 10      # Price Y

# Grid
x = np.linspace(0, 20, 100)

# Budget Line: 10x + 10y = 100 -> y = 10 - x
budget_y = (M - px * x) / py

# Indifference Curve (Cobb-Douglas U = xy)
# Optimal is x=5, y=5 -> U = 25
# y = 25/x
u_optimal = 25
ic_y = u_optimal / x

# Sub-optimal Indifference Curve (U = 16)
ic_subopt = 16 / x

plt.figure(figsize=(8, 8))

# Plot Lines
plt.plot(x, budget_y, label='Budget Line', color='red', linewidth=2)
plt.plot(x, ic_y, label='Optimal IC (U=25)', color='green', linewidth=2)
plt.plot(x, ic_subopt, label='Sub-optimal IC (U=16)', color='green', linestyle='--', alpha=0.5)

# Tangency Point
plt.plot(5, 5, 'ro', label='Optimal Bundle (5, 5)')

# Formatting
plt.title('Optimization: Tangency')
plt.xlabel('Good X')
plt.ylabel('Good Y')
plt.xlim(0, 12)
plt.ylim(0, 12)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

**Graph Visualization:**
*   The **Red Line** is the Budget Constraint (The Fence).
*   The **Green Curve** is the Indifference Curve (The Mountain).
*   The **Dot** is the highest point on the mountain you can reach without jumping the fence.
*   At the dot, the slopes are identical.

#### **(4) Key Takeaways**
*   **Interior Solution:** You buy some of both. Condition: $MRS = Price Ratio$.
*   **Corner Solution:** You buy only one good. Occurs with Perfect Substitutes (e.g., if Coke is cheaper than Pepsi, you buy 0 Pepsi).
*   **The Lagrangian Method:** This is the math "power tool" used in your PDFs.
    *   $\mathcal{L} = U(x,y) + \lambda (M - p_x x - p_y y)$
    *   $\lambda$ (Lambda) represents the **Marginal Utility of Income**. It tells you: "If I gave you \$1 more, how much would your utility go up?"

#### **(5) Worked Example (Cobb-Douglas)**
**Scenario:** Ash has $M = \$30$.
*   $P_B = 10$, $P_F = 5$.
*   Utility: $U(B, F) = B \cdot F$.

**Step 1: Find MRS.**
*   $MU_B = F$ (derivative of $B \cdot F$ with respect to $B$).
*   $MU_F = B$.
*   $MRS = \frac{MU_B}{MU_F} = \frac{F}{B}$.

**Step 2: Set MRS equal to Price Ratio.**

### **Chapter 4: Demand & Elasticity**

#### **(1) Intuition First**
Its a rainy Tuesday at **The Eco-Nom-Nom Food Truck**. To boost sales, we decide to put **Burgers on sale** (Price drops from \$10 to \$5).

How does Ash respond? He buys *more* Burgers. But **why**? Economics breaks this decision into two distinct psychological steps:

1.  **The Substitution Effect ("The Swap"):** Burgers are now cheaper relative to Fries. Even if Ashs actual purchasing power stayed the same, he would naturally swap expensive Fries for cheap Burgers. *This effect is always negative* (Price down $\rightarrow$ Quantity up).
2.  **The Income Effect ("The Bonus"):** Because Burgers are cheaper, Ash has "extra money" left over. It feels like he got a raise. He uses this "extra" wealth to buy more of *everything* he likes.

**The Twist:** What if Ash was eating instant noodles (an **Inferior Good**)?
*   **Substitution:** Noodles get cheaper $\rightarrow$ Buy more noodles.
*   **Income:** Ash feels richer $\rightarrow$ "I'm too rich for noodles now!" $\rightarrow$ Buy *fewer* noodles.
*   The two effects fight each other!

#### **(2) Formal Definition (Translated)**

*   **Demand Function ($x(p, M)$):** A mathematical recipe telling us how much $x$ Ash buys given the price $p$ and his income $M$.
*   **Normal Good:** When Income ($M$) goes up, Demand goes up. (Steak, iPhones).
*   **Inferior Good:** When Income ($M$) goes up, Demand goes down. (Instant noodles, bus tickets).
*   **Giffen Good:** A weird theoretical monster. An inferior good where the Income Effect is *so strong* it overpowers the Substitution Effect. Price goes up $\rightarrow$ You feel poorer $\rightarrow$ You panic and buy *more* of the cheap survival food.

**The Slutsky Equation (The Decomposition):**
$$ \text{Total Change} = \text{Substitution Effect} + \text{Income Effect} $$

$$\frac{\Delta x}{\Delta p} = \underbrace{\left( \frac{\Delta x}{\Delta p} \right)_{comp}}_{\text{Substitution}} - \underbrace{x \cdot \left( \frac{\Delta x}{\Delta M} \right)}_{\text{Income}}$$

#### **(3) Graphs & Relationships**

We will visualize the **Substitution and Income Effects** using the "Pivot-Shift" method.

1.  **Original Choice (A):** High Price.
2.  **Intermediate Choice (B):** We create a "fake" budget line. It has the *new* prices (slope) but enough income to buy the *old* bundle. This isolates the "Swap."
3.  **Final Choice (C):** New prices, actual income. This captures the "Richness."

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
M = 100
px_old = 10
px_new = 5
py = 10

# Utility: U = xy (Cobb-Douglas)
# Demand X = M / (2*px)
# Demand Y = M / (2*py)

# 1. Point A (Original)
xA = M / (2 * px_old) # 5
yA = M / (2 * py)     # 5
U_A = xA * yA         # 25

# 2. Point C (Final)
xC = M / (2 * px_new) # 10
yC = M / (2 * py)     # 5
U_C = xC * yC         # 50

# 3. Point B (Intermediate - Slutsky Compensation)
# We need enough income (M') to afford bundle A at NEW prices
M_prime = px_new * xA + py * yA  # 5*5 + 10*5 = 75
xB = M_prime / (2 * px_new)      # 75 / 10 = 7.5
yB = M_prime / (2 * py)          # 75 / 20 = 3.75

# Plotting
x = np.linspace(0, 20, 100)

# Budget Lines
bl_old = (M - px_old * x) / py
bl_new = (M - px_new * x) / py
bl_fake = (M_prime - px_new * x) / py

plt.figure(figsize=(10, 8))

# Draw Budget Lines
plt.plot(x, bl_old, 'r--', label='Old Budget (High Price)', alpha=0.5)
plt.plot(x, bl_new, 'b-', label='New Budget (Low Price)', linewidth=2)
plt.plot(x, bl_fake, 'g--', label='Fake Budget (Pivot)', alpha=0.6)

# Draw Points
plt.plot(xA, yA, 'ro', markersize=8, label='A (Original)')
plt.plot(xB, yB, 'go', markersize=8, label='B (Sub Effect)')
plt.plot(xC, yC, 'bo', markersize=8, label='C (Total Effect)')

# Draw Arrows
plt.arrow(xA, yA+0.5, xB-xA, -1.75, length_includes_head=True, head_width=0.3, color='green')
plt.text(6, 6, "Sub Effect (A->B)", color='green')

plt.arrow(xB, yB+0.5, xC-xB, 1.25, length_includes_head=True, head_width=0.3, color='blue')
plt.text(8.5, 5.5, "Income Effect (B->C)", color='blue')

# Formatting
plt.title('Decomposing Price Changes (Slutsky Method)')
plt.xlabel('Burgers')
plt.ylabel('Fries')
plt.xlim(0, 15)
plt.ylim(0, 12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

**Understanding the Graph:**
*   **A to B (Substitution):** The Green dashed line represents the "New Relative Price" but with "Old Purchasing Power." Ash moves from 5 Burgers to 7.5 Burgers just because they are cheaper relative to Fries.
*   **B to C (Income):** The Blue line represents the actual new reality. Ash effectively has \$25 extra dollars. He uses it to buy *even more* Burgers (7.5 to 10).

#### **(4) Key Takeaways**
*   **Law of Demand:** Demand curves slope down (usually).
*   **Income Effect:** Can be positive (Normal) or negative (Inferior).
*   **Substitution Effect:** Always negative (Price $\downarrow$ Quantity $\uparrow$).
*   **Price Elasticity of Demand ($\epsilon$):** $\% \Delta Q / \% \Delta P$.
    *   $|\epsilon| > 1$: Elastic (Luxury goods, many substitutes).
    *   $|\epsilon| < 1$: Inelastic (Necessities, few substitutes).

#### **(5) Worked Example (Calculating Elasticity)**
**Scenario:**
*   At Price $P = \$10$, Ash buys $Q = 5$ Burgers.
*   At Price $P = \$9$, Ash buys $Q = 6$ Burgers.

**Step 1: Calculate % Changes.**
*   $\% \Delta P = (9 - 10) / 10 = -0.10$ (-10%).
*   $\% \Delta Q = (6 - 5) / 5 = +0.20$ (+20%).

**Step 2: Calculate Elasticity.**
$$ \epsilon = \frac{+20\%}{-10\%} = -2 $$

**Interpretation:**
For every 1% price drop, Ash buys 2% more burgers. Demand is **Elastic**. The truck owner should lower prices to increase total revenue!

### **Chapter 5: Decisions over Time & Risk**

#### **(1) Intuition First**
Business at **The Eco-Nom-Nom Food Truck** is booming. Now Ash (our loyal customer) and you (the owner) face two new types of choices:
1.  **Time:** "Do I eat a cheap burger today, or save my money to buy a fancy steak next week?"
2.  **Risk:** "Do I buy the 'Mystery Meat Special' (50% chance it's delicious, 50% chance it's awful)?"

**Time:** Money today is worth more than money tomorrow. Why? Because you can invest money today and earn interest. To compare money across time, we must "discount" the future back to the present.

**Risk:** Most people dislike risk (Risk Aversion). If I offer you \$50 for sure OR a coin flip for \$100/zero, most people take the \$50. Why? Because the pain of losing is stronger than the joy of winning (Diminishing Marginal Utility).

#### **(2) Formal Definition (Translated)**

*   **Present Value (PV):** The value of future money *right now*.
    $$PV = \frac{Future Value}{(1+r)^t}$$
    where $r$ is the interest rate and $t$ is time.

*   **Intertemporal Budget Constraint:**
    $$c_1 + \frac{c_2}{1+r} = m_1 + \frac{m_2}{1+r}$$
    *   *Translation:* The Present Value of your consumption ($c$) must equal the Present Value of your income ($m$).

*   **Expected Utility ($EU$):** We don't average the *money*; we average the *happiness* (utility) from the money.
    $$EU = \pi_1 U(w_1) + \pi_2 U(w_2)$$
    where $\pi$ is probability and $w$ is wealth.

#### **(3) Graphs & Relationships**

**Risk Aversion Graph:**
A concave utility function means the curve flattens out. The utility of the "average wealth" is higher than the "average utility" of the gamble.

```python
import matplotlib.pyplot as plt
import numpy as np

# Utility Function: U = sqrt(w) (Risk Averse)
w = np.linspace(0, 100, 100)
u = np.sqrt(w)

# Gamble: 50% chance of 0, 50% chance of 100
# Expected Wealth = 50
w_low = 0
w_high = 100
w_avg = 50

u_low = 0
u_high = 10  # sqrt(100)
eu = 0.5 * u_low + 0.5 * u_high  # EU = 5

u_sure_thing = np.sqrt(50) # 7.07

plt.figure(figsize=(8, 6))
plt.plot(w, u, label='Utility U(w) = sqrt(w)', linewidth=2)

# Chord representing expected utility
plt.plot([0, 100], [0, 10], 'r--', label='Expected Utility Line')

# Points
plt.plot(50, 5, 'ro', label='EU of Gamble (5)')
plt.plot(50, 7.07, 'go', label='U of Sure Thing (7.07)')

plt.vlines(50, 0, 7.07, linestyle=':', color='gray')
plt.text(52, 2, 'Risk Premium Gap', fontsize=10)

plt.title('Risk Aversion: Concave Utility')
plt.xlabel('Wealth ($)')
plt.ylabel('Utility')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

**Graph Insight:**
*   **Green Dot:** Happiness from having \$50 for sure.
*   **Red Dot:** Average happiness from the gamble ($0 or $100).
*   The Green Dot is higher. The consumer would pay money (a **Risk Premium**) to avoid the gamble.

#### **(4) Key Takeaways**
*   **Time Value:** Money shrinks as it moves into the future ($1/(1+r)$).
*   **Risk Aversion:** Comes from **Diminishing Marginal Utility** ($U'' < 0$). The first million dollars makes you happier than the second million.
*   **Certainty Equivalent:** The guaranteed cash amount that makes you just as happy as the risky gamble.

#### **(5) Worked Example (Intertemporal Choice)**
**Scenario:** Ash earns \$100 today ($t=1$) and \$110 next year ($t=2$). Interest rate $r = 10\%$ ($0.1$).
He wants to consume the *same amount* in both periods ($c_1 = c_2 = c$).

**Step 1: Calculate PV of Income.**
$$PV_{Inc} = 100 + \frac{110}{1.1} = 100 + 100 = 200$$

**Step 2: Set up PV of Consumption.**
$$PV_{Cons} = c + \frac{c}{1.1} = c(1 + 0.909) = 1.909c$$

**Step 3: Equate and Solve.**
$$1.909c = 200 \Rightarrow c = \frac{200}{1.909} \approx 104.76$$

**Result:** Ash spends \$104.76 today (borrowing \$4.76) and \$104.76 next year. He smooths his consumption.

We are leaving the sidewalk (Consumer Theory) and stepping inside **The Eco-Nom-Nom Food Truck**. You are now the Manager. Your goal is to produce **Burgers ($Q$)**.

To make burgers, you need two things:
1.  **Labor ($L$):** Cooks and cashiers.
2.  **Capital ($K$):** Grills, fryers, and the truck itself.

**The "Recipe" (Technology):**
Just like a chemical equation, you combine $K$ and $L$ to get $Q$.
*   **Marginal Product (The "Extra" Effect):** If you hire one more cook, how many extra burgers can you make?
    *   *The Law of Diminishing Returns:* Imagine a small truck with one grill.
        *   1 Cook: 10 burgers/hour.
        *   2 Cooks: 18 burgers/hour (They help each other).
        *   3 Cooks: 20 burgers/hour (Getting crowded).
        *   10 Cooks: 0 burgers/hour (They can't move!).
    *   *Key Insight:* Adding more Labor *eventually* becomes less useful if Capital is fixed.

**Scaling Up:**
What if we double **everything**? We buy a second truck, a second grill, and hire a second team.
*   If Output Doubles ($2Q$): **Constant Returns to Scale (CRS)**. (Standard franchise model).
*   If Output *More* than Doubles ($>2Q$): **Increasing Returns to Scale (IRS)**. (Specialization/Efficiency).
*   If Output *Less* than Doubles ($<2Q$): **Decreasing Returns to Scale (DRS)**. (Management issues).

#### **(2) Formal Definition (Translated)**

*   **Production Function $Q = F(K, L)$:** The mathematical rule turning inputs into outputs.
*   **Isoquant:** A curve showing all combinations of $K$ and $L$ that produce the **same quantity** of output. (This is exactly like an Indifference Curve, but for producers).
*   **Marginal Product of Labor ($MP_L$):** $\frac{\partial Q}{\partial L}$. The slope of the total product curve.
*   **Marginal Rate of Technical Substitution ($MRTS$):** The slope of the Isoquant.
    $$MRTS = \frac{MP_L}{MP_K}$$
    *   *Translation:* How many machines ($K$) can I get rid of if I hire one more person ($L$), while keeping production exactly the same?

#### **(3) Graphs & Relationships**

We will visualize **Isoquants** (Technology Maps).

```python
import matplotlib.pyplot as plt
import numpy as np

# Grid
L = np.linspace(0.1, 10, 100)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Cobb-Douglas (Smooth Substitution)
# Q = K * L -> K = Q / L
# This allows substitution but requires both inputs
for q in [5, 10, 20]:
    axes[0].plot(L, q/L, label=f'Output={q}', linewidth=2)
axes[0].set_title("Cobb-Douglas (Standard Kitchen)\nSmooth Substitution")
axes[0].set_xlabel("Labor (Cooks)")
axes[0].set_ylabel("Capital (Grills)")
axes[0].set_xlim(0, 10)
axes[0].set_ylim(0, 10)
axes[0].grid(True, alpha=0.3)

# 2. Perfect Complements (Fixed Proportions)
# Q = min(L, K) -> L-shaped
# One driver needs exactly one truck.
for q in [2, 4, 6]:
    axes[1].vlines(q, q, 10, colors='blue', linewidth=2)
    axes[1].hlines(q, q, 10, colors='blue', linewidth=2, label=f'Q={q}')
axes[1].set_title("Perfect Complements (Leontief)\n1 Driver : 1 Truck")
axes[1].set_xlabel("Drivers")
axes[1].set_ylabel("Trucks")
axes[1].set_xlim(0, 10)
axes[1].set_ylim(0, 10)
axes[1].grid(True, alpha=0.3)

# 3. Perfect Substitutes
# Q = L + K -> K = Q - L
# Robots vs Humans. They do the exact same job.
for q in [5, 8, 10]:
    axes[2].plot(L, q - L, label=f'Q={q}', linewidth=2)
axes[2].set_title("Perfect Substitutes\nRobots vs Humans")
axes[2].set_xlabel("Humans")
axes[2].set_ylabel("Robots")
axes[2].set_xlim(0, 10)
axes[2].set_ylim(0, 10)
axes[2].grid(True, alpha=0.3)

plt.show()
```

**Graph Breakdown:**
*   **Standard Kitchen (Left):** You can substitute. If a grill breaks ($K \downarrow$), you can maintain output by hiring more cooks ($L \uparrow$) to work faster on the remaining grill.
*   **1 Driver : 1 Truck (Middle):** No substitution. Adding a 2nd driver to 1 truck yields zero extra output.
*   **Robots vs Humans (Right):** Complete substitution. The slope is constant ($MRTS$ is constant).

#### **(4) Key Takeaways**
*   **Diminishing Marginal Product** happens in the **Short Run** (when Capital is fixed).
*   **Returns to Scale** is a **Long Run** concept (when you can change *everything*).
*   **MRTS** measures the trade-off between inputs. It usually diminishes (it's harder to replace machines with people when you have almost no machines left).

#### **(5) Worked Example (Returns to Scale)**
**Scenario:** Production Function $Q = 10 K^{0.5} L^{0.5}$.
Current setup: 4 Grills, 4 Cooks.
*   $Q_1 = 10 (4)^{0.5} (4)^{0.5} = 10(2)(2) = 40$ Burgers.

**Experiment:** Double both inputs to 8.
*   $Q_2 = 10 (8)^{0.5} (8)^{0.5} = 10(2.82)(2.82) = 80$ Burgers.

**Result:**
*   Inputs doubled ($2x$).
*   Output doubled ($40 \rightarrow 80$).
*   **Conclusion:** Constant Returns to Scale (CRS).

**Shortcut:** Add the exponents. $0.5 + 0.5 = 1$.
*   Sum $= 1 \rightarrow$ CRS.
*   Sum $> 1 \rightarrow$ IRS.
*   Sum $< 1 \rightarrow$ DRS.

### **Chapter 7: Costs & Supply (The Bottom Line)**

#### **(1) Intuition First**
Back at **The Eco-Nom-Nom Food Truck**. In Chapter 6, we figured out the "recipe" (Technology) to make burgers using Cooks ($L$) and Grills ($K$). Now, we have to pay for them.

*   **Cost Minimization:** Suppose we need to produce 100 burgers for a catering event. We could use 100 cooks on 1 grill (very crowded, expensive wages) or 1 cook on 100 grills (expensive equipment rent).
    *   We want the **cheapest combination** of cooks and grills that gets the job done.
    *   This is exactly like the Consumer's problem, but reversed. Instead of maximizing happiness for a fixed budget, we **minimize the budget** for a fixed target output.

*   **Marginal Cost (MC):** Once we are running efficiently, we ask: "How much does it cost to make **one more** burger?"
    *   If the kitchen is empty, it's cheap (just ingredients).
    *   If the kitchen is crowded, it's expensive (overtime pay, waiting for the grill).
    *   This drives our **Supply Decision**. We keep cooking as long as the Price someone pays is higher than the Cost of making that specific burger ($P \geq MC$).

#### **(2) Formal Definition (Translated)**

*   **Isocost Line:** A line showing all combinations of Labor ($L$) and Capital ($K$) that cost the same amount.
    $$Cost = wL + rK$$
    (Slope $= -w/r$).

*   **Cost Minimization Condition:**
    $$\frac{MP_L}{MP_K} = \frac{w}{r}$$
    *   *Translation:* The ratio of productivity must equal the ratio of costs. If cooks are twice as productive as machines but cost 4x as much, fire cooks and buy machines.

*   **Cost Curves:**
    *   **Total Cost ($TC$):** Fixed Cost ($F$) + Variable Cost ($VC$).
    *   **Average Total Cost ($ATC$):** Total Cost per burger ($TC/Q$). "Breaking Even."
    *   **Marginal Cost ($MC$):** Cost of the *next* burger ($\partial TC / \partial Q$). "Profit Maximizing."

#### **(3) Graphs & Relationships**

We will visualize the **Cost Curves**. These are the most famous shapes in economics (the "U-shapes").

```python
import matplotlib.pyplot as plt
import numpy as np

# Define Cost Function: C(q) = 100 + q^2
# Fixed Cost = 100
# Variable Cost = q^2
q = np.linspace(1, 20, 100)

TC = 100 + q**2
AFC = 100 / q          # Average Fixed Cost
AVC = q                # Average Variable Cost (q^2 / q)
ATC = (100 / q) + q    # Average Total Cost
MC = 2 * q             # Marginal Cost (derivative of q^2)

plt.figure(figsize=(10, 6))

# Plot Curves
plt.plot(q, MC, label='Marginal Cost (MC)', color='red', linewidth=2)
plt.plot(q, ATC, label='Average Total Cost (ATC)', color='blue', linewidth=2)
plt.plot(q, AVC, label='Average Variable Cost (AVC)', color='green', linestyle='--')
plt.plot(q, AFC, label='Average Fixed Cost (AFC)', color='gray', linestyle=':')

# Formatting
plt.title('The Family of Cost Curves')
plt.xlabel('Quantity of Burgers (Q)')
plt.ylabel('Cost ($)')
plt.ylim(0, 50)
plt.xlim(0, 20)
plt.grid(True, alpha=0.3)
plt.legend()

# Highlight intersections
# MC cuts ATC at its minimum
min_atc_q = 10 # sqrt(100)
min_atc_val = 20
plt.plot(min_atc_q, min_atc_val, 'ko') 
plt.text(min_atc_q, min_atc_val+2, 'Break-Even Point', fontsize=10)

plt.show()
```

**Graph Breakdown:**
1.  **MC (Red Line):** Goes up. It eventually gets harder/more expensive to squeeze more output out of a fixed kitchen.
2.  **ATC (Blue U-Shape):**
    *   At low $Q$, it falls (spreading the Fixed Cost of the truck over more burgers).
    *   At high $Q$, it rises (overcrowding/diminishing returns kick in).
3.  **Intersection:** The MC line cuts through the **bottom** of the ATC curve.
    *   *Analogy:* If your next exam score (Marginal) is higher than your GPA (Average), your GPA goes up. If it's lower, your GPA goes down. They cross exactly where the average stops falling and starts rising.

#### **(4) Key Takeaways**
*   **Supply Rule:** A competitive firm produces where $P = MC$.
*   **Shutdown Rule:** If Price drops below **AVC** (Average Variable Cost), shut down immediately. You aren't even covering the ingredients!
*   **Exit Rule:** If Price is below **ATC** (Average Total Cost), you are losing money. In the Long Run, exit the industry.

#### **(5) Worked Example (Cost Minimization)**
**Scenario:**
*   Target Output: $Q = 100$ Burgers.
*   Technology: $Q = \sqrt{K \cdot L}$.
*   Wage ($w$) = \$10. Rental Rate ($r$) = \$40.

**Step 1: Find MRTS.**
*   $MP_L = 0.5 \sqrt{K/L}$.
*   $MP_K = 0.5 \sqrt{L/K}$.
*   $MRTS = K/L$.

**Step 2: Set MRTS = Price Ratio.**
*   $K/L = w/r = 10/40 = 1/4$.
*   $L = 4K$.
*   *Intuition:* Labor is 4x cheaper than Capital, so we use 4x as much Labor.

**Step 3: Plug into Production Function.**
*   $100 = \sqrt{K \cdot (4K)} = \sqrt{4K^2} = 2K$.
*   $100 = 2K \Rightarrow K = 50$.
*   $L = 4(50) = 200$.

**Step 4: Calculate Total Cost.**
*   $Cost = 10(200) + 40(50) = 2000 + 2000 = \$4000$.

### **Chapter 8: Monopoly Power (The Only Truck in Town)**

#### **(1) Intuition First**
Great news for **The Eco-Nom-Nom Food Truck**: We have secured an exclusive contract to be the *only* food vendor inside a music festival. There is no competition. We are a **Monopoly**.

In the previous chapters (Perfect Competition), we were "Price Takers." The market set the price ($\$10$), and we just decided how many burgers to flip. If we charged $\$10.01$, we sold zero.

Now, we are "Price Makers." The demand curve for our burgers *is* the market demand curve. It slopes down.
*   **The Trade-off:** We can charge $\$20$ and sell to only a few rich foodies, or charge $\$5$ and feed the whole crowd.
*   **The Poison Effect:** To sell *one more* burger, we have to lower the price. Butand this is the keywe usually have to lower the price for *everyone*, not just the new customer. This "loss" on existing customers makes our **Marginal Revenue (MR)** lower than the Price.

#### **(2) Formal Definition (Translated)**

*   **Marginal Revenue (MR):** The extra money we get from selling one more unit.
    $$MR = P + Q \frac{\Delta P}{\Delta Q}$$
    *   *Translation:* (Price of the new unit) minus (The discount we gave to everyone else $\times$ quantity sold).
    *   *Linear Shortcut:* If Demand is a straight line ($P = a - bQ$), Marginal Revenue is a straight line with **twice the slope** ($MR = a - 2bQ$).

*   **Profit Maximization Rule:**
    $$MR = MC$$
    *   We keep expanding until the revenue from the last burger exactly equals the cost to make it. Since $P > MR$ and $MR = MC$, it follows that **Price > Marginal Cost**. We charge a markup!

*   **The Lerner Index (Market Power):**
    $$L = \frac{P - MC}{P} = -\frac{1}{\epsilon}$$
    *   The more inelastic ($\epsilon$) the customers are (i.e., the more desperate/addicted they are), the higher the markup we can charge.

#### **(3) Graphs & Relationships**

We will visualize the **Monopoly Equilibrium** and the **Deadweight Loss** (the lost value because prices are too high).

```python
import matplotlib.pyplot as plt
import numpy as np

# Linear Demand: P = 10 - Q
# Marginal Revenue: MR = 10 - 2Q
# Marginal Cost: Constant MC = 2

Q = np.linspace(0, 10, 100)
P_demand = 10 - Q
MR = 10 - 2 * Q
MC = np.full_like(Q, 2)

plt.figure(figsize=(8, 6))

# Plot Curves
plt.plot(Q, P_demand, label='Demand (AR)', color='blue', linewidth=2)
plt.plot(Q, MR, label='Marginal Revenue (MR)', color='green', linestyle='--')
plt.plot(Q, MC, label='Marginal Cost (MC)', color='red', linewidth=2)

# Calculation for Equilibrium
# MR = MC -> 10 - 2Q = 2 -> 2Q = 8 -> Q* = 4
Q_monop = 4
P_monop = 10 - 4 # = 6

# Calculation for Social Optimum (Perfect Competition)
# P = MC -> 10 - Q = 2 -> Q_soc = 8
Q_soc = 8
P_soc = 2

# Highlight Points
plt.plot(Q_monop, P_monop, 'ko')
plt.text(Q_monop+0.2, P_monop, f'Monopoly Price (${P_monop})', fontsize=10)

plt.plot(Q_monop, 2, 'ko') # MR=MC intersection

# Shade Deadweight Loss
# Triangle between Q_monop and Q_soc, below Demand and above MC
plt.fill_between(Q, P_demand, 2, where=(Q >= Q_monop) & (Q <= Q_soc), 
                 color='gray', alpha=0.3, label='Deadweight Loss')

# Formatting
plt.title('Monopoly Pricing & Deadweight Loss')
plt.xlabel('Quantity (Q)')
plt.ylabel('Price ($)')
plt.ylim(0, 11)
plt.xlim(0, 10)
plt.axhline(0, color='black')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

**Graph Breakdown:**
1.  **Monopoly Quantity ($Q_M=4$):** Found where Green ($MR$) hits Red ($MC$).
2.  **Monopoly Price ($P_M=6$):** Go **UP** from the quantity to the Blue line (Demand). That's what people will pay.
3.  **Social Optimum ($Q_{soc}=8$):** Where Supply ($MC$) hits Demand.
4.  **Deadweight Loss (Grey Triangle):** These are the hungry people willing to pay \$5, \$4, or \$3 for a burger. Since it costs us only \$2 to make, society *should* feed them. But the Monopolist won't, because lowering the price to \$5 spoils the profits from the rich customers paying \$6.

#### **(4) Price Discrimination (The "Hacks")**
Monopolists hate leaving money on the table (Deadweight Loss). They try to charge different prices to different people to capture that surplus.

*   **1st Degree (Perfect):** The Mind Reader. You charge every single person exactly their maximum willingness to pay.
    *   *Result:* Zero Deadweight Loss, but the Monopolist takes *all* the surplus.
*   **2nd Degree (Self-Selection):** The "Menu". You can't read minds, so you offer quantity discounts.
    *   *Example:* Single Burger \$10. "Double-Double" \$15. High-appetite people reveal themselves by choosing the big meal.
*   **3rd Degree (Segmentation):** The ID Card. You separate markets by observable traits.
    *   *Example:* Students get 10% off. Why? Students usually have lower income (more elastic demand).

#### **(5) Worked Example (3rd Degree Discrimination)**
**Scenario:** Our truck has two lines.
*   **Adults (Group A):** Inverse Demand $P_A = 100 - Q_A$.
*   **Students (Group S):** Inverse Demand $P_S = 50 - Q_S$.
*   **Cost:** $MC = 10$ (Constant).

**Strategy:** Treat them as two separate monopolies. Set $MR = MC$ for each.

**Group A (Adults):**
*   $P_A = 100 - Q_A \Rightarrow MR_A = 100 - 2Q_A$.
*   Set $MR_A = MC$: $100 - 2Q_A = 10 \Rightarrow 2Q_A = 90 \Rightarrow Q_A = 45$.
*   Price: $P_A = 100 - 45 = \$55$.

**Group S (Students):**
*   $P_S = 50 - Q_S \Rightarrow MR_S = 50 - 2Q_S$.
*   Set $MR_S = MC$: $50 - 2Q_S = 10 \Rightarrow 2Q_S = 40 \Rightarrow Q_S = 20$.
*   Price: $P_S = 50 - 20 = \$30$.

**Result:** We charge Adults \$55 and Students \$30. We make more money than if we forced a single price.

### **Chapter 9: Game Theory & Oligopoly (Food Truck Wars)**

#### **(1) Intuition First**
The glory days of the Monopoly are over. A rival truck, **"The Burger Bros,"** has parked right across the street.

We are now in an **Oligopoly** (Few Sellers). We can no longer just look at the Demand Curve and pick a price. Now, our profits depend on **what the other truck does**.
*   **The Scenario:**
    *   If we *both* set high prices (\$10), we split the rich customers and make good money (**Collusion/Cartel**).
    *   If *I* cut my price to \$9 and *they* stay at \$10, I steal all their customers and make a fortune.
    *   If *they* cut to \$8 to punish me, I have to cut to \$7.
    *   Eventually, we both end up selling burgers for \$2 (Marginal Cost), making zero profit. This is the **Prisoner's Dilemma**.

We analyze this "Strategic Interdependence" using **Game Theory**.

#### **(2) Formal Definition (Translated)**

*   **Nash Equilibrium (NE):** A situation where **no player regrets their choice** given what the other player did.
    *   *The Test:* "If I knew you were going to do $X$, would I still do $Y$?" If the answer is "Yes" for both players, it's a Nash Equilibrium.
*   **Best Response Function ($BR$):** A formula telling you your optimal move for every possible move the opponent might make.
    *   *Mathematical notation:* $q_1^* = BR_1(q_2)$.

**The Three Main Models of Battle:**
1.  **Cournot (Quantity War):** We both decide how much prep (inventory) to do in the morning *at the same time*.
    *   *Result:* Prices end up lower than Monopoly, but higher than Competitive.
2.  **Bertrand (Price War):** We both write our prices on a chalkboard *at the same time*.
    *   *Result:* If products are identical, we undercut each other until Price = Marginal Cost (Zero Profit).
3.  **Stackelberg (Leader-Follower):** I open my truck an hour before you. I commit to a quantity first. You see my output, then decide yours.
    *   *Result:* I (the Leader) bully you. I produce more, forcing you (the Follower) to retreat and produce less.

#### **(3) Graphs & Relationships**

We will visualize **Cournot Reaction Functions**.

```python
import matplotlib.pyplot as plt
import numpy as np

# Market Demand: P = 100 - Q (where Q = q1 + q2)
# MC = 10
# Profit 1 = (100 - q1 - q2 - 10) * q1
# FOC for Firm 1: 90 - 2q1 - q2 = 0 -> q1 = (90 - q2) / 2

q2 = np.linspace(0, 100, 100)

# Reaction Function 1: q1 = 45 - 0.5*q2
rf1 = 45 - 0.5 * q2

# Reaction Function 2: q2 = 45 - 0.5*q1 -> Invert for plotting: q1 = 90 - 2*q2
# To plot q2 on y-axis, we use q2 = 45 - 0.5*q1
q1_axis = np.linspace(0, 100, 100)
rf2_plot = 45 - 0.5 * q1_axis

plt.figure(figsize=(8, 8))

# Plot RF1 (Firm 1's best response to Firm 2)
# Note: Usually we plot q1 on x, q2 on y.
# If q2=0, q1=45. If q2=90, q1=0.
plt.plot(rf1, q2, label="Firm 1's Reaction Curve", color='blue', linewidth=2)

# Plot RF2 (Firm 2's best response to Firm 1)
# If q1=0, q2=45. If q1=90, q2=0.
plt.plot(q1_axis, rf2_plot, label="Firm 2's Reaction Curve", color='red', linewidth=2)

# Intersection (Nash Equilibrium)
# q = 45 - 0.5q -> 1.5q = 45 -> q = 30
plt.plot(30, 30, 'ko', markersize=10, label='Cournot Nash Equilibrium (30, 30)')

# Formatting
plt.title('Cournot Competition: Reaction Curves')
plt.xlabel("Firm 1 Quantity (q1)")
plt.ylabel("Firm 2 Quantity (q2)")
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

**Graph Breakdown:**
*   **Blue Line:** If Firm 2 brings 0 burgers (y-axis), Firm 1 acts like a Monopoly and brings 45 (x-axis). As Firm 2 brings more, Firm 1 retreats (brings fewer).
*   **Red Line:** Symmetry. If Firm 1 brings 0, Firm 2 brings 45.
*   **Intersection:** The only point where *both* are happy. Neither wants to move. $(30, 30)$.

#### **(4) Key Takeaways**
*   **Cournot:** Output is **Strategic Substitutes**. If you do more, I do less.
*   **Bertrand:** Prices are **Strategic Complements**. If you cut price, I must cut price.
*   **First Mover Advantage (Stackelberg):** In quantity competition, moving first allows you to capture market share.
*   **Second Mover Advantage:** In price competition, moving second allows you to undercut the other guy slightly.

#### **(5) Worked Example (Cournot Calculation)**
**Scenario:**
*   Demand: $P = 120 - Q$ where $Q = q_1 + q_2$.
*   Marginal Cost ($c$) = 0 (for simplicity).

**Step 1: Firm 1's Problem.**
*   Profit $\pi_1 = P \cdot q_1 = (120 - q_1 - q_2)q_1$.
*   $\pi_1 = 120q_1 - q_1^2 - q_2q_1$.

**Step 2: Marginal Revenue = Marginal Cost.**
*   Take derivative with respect to $q_1$ (treat $q_2$ as a constant number).
*   $MR_1 = 120 - 2q_1 - q_2$.
*   Set $MR_1 = MC = 0$.
*   $120 - 2q_1 - q_2 = 0 \Rightarrow 2q_1 = 120 - q_2$.
*   **Reaction Function 1:** $q_1^* = 60 - 0.5q_2$.

**Step 3: Symmetry.**
*   Since Firm 2 is identical, $q_2^* = 60 - 0.5q_1$.

**Step 4: Solve the System.**
*   Substitute (2) into (1):
    $q_1 = 60 - 0.5(60 - 0.5q_1)$.
    $q_1 = 60 - 30 + 0.25q_1$.
    $0.75q_1 = 30 \Rightarrow q_1 = 40$.
*   Since symmetric, $q_2 = 40$.

**Result:** Total $Q = 80$. Price $P = 120 - 80 = 40$. Profit per firm = $40 \times 40 = 1600$.

### **Chapter 10: General Equilibrium (The Whole Economy)**

#### **(1) Intuition First**
So far, weve analyzed **The Eco-Nom-Nom Food Truck** in isolation (Partial Equilibrium). We looked at the burger market assuming the price of everything else (rent, gas, wages) was fixed.

**General Equilibrium** looks at the *entire* city at once.
Imagine Ash (Consumer A) and Misty (Consumer B) are stuck on an island. They have a pile of Burgers and a pile of Fries. No production, just trading.
*   **The Scenario:** Ash has 10 Burgers and 0 Fries. Misty has 0 Burgers and 10 Fries.
*   **The Trade:** Ash is desperate for Fries. Hed happily give up 2 Burgers for 1 Fry. Misty is desperate for Burgers. Shed give 2 Fries for 1 Burger.
*   **The Win-Win:** They trade until they reach a point where neither wants to trade anymore.
*   **Pareto Efficiency:** A situation where you cannot make Ash happier without making Misty sadder. If there is a deal on the table that helps *both* (or helps one and hurts no one), the current situation is **not** efficient.

#### **(2) Formal Definition (Translated)**

*   **Edgeworth Box:** A graphical tool representing the entire economy.
    *   Width = Total Burgers in existence.
    *   Height = Total Fries in existence.
    *   Ash is at the **Bottom-Left** corner.
    *   Misty is at the **Top-Right** corner (upside down).

*   **Contract Curve:** The set of all Pareto Efficient points.
    *   Mathematically: Where Ashs Indifference Curve is tangent to Mistys Indifference Curve.
    *   Condition: $MRS_{Ash} = MRS_{Misty}$.

*   **First Fundamental Theorem of Welfare Economics:**
    If markets are competitive (no monopolies) and complete, the equilibrium is **Pareto Efficient**.
    *   *Translation:* The Invisible Hand works. You don't need a dictator to tell Ash and Misty how to swap lunch. The price system ($P_B, P_F$) will guide them to the perfect allocation automatically.

#### **(3) Graphs & Relationships**

We will build an **Edgeworth Box**.

```python
import matplotlib.pyplot as plt
import numpy as np

# Total Endowments
Total_X = 10
Total_Y = 10

# Grid for Ash (Origin 0,0)
x_A = np.linspace(0.1, 9.9, 100)

# Contract Curve:
# Assume identical Cobb-Douglas U = xy for both.
# MRS_A = yA/xA. MRS_B = yB/xB.
# yB = 10 - yA, xB = 10 - xA.
# Equate MRS: yA/xA = (10-yA)/(10-xA) -> yA = xA (Straight diagonal line)
y_contract = x_A

# Indifference Curves
# Ash U=25 -> y = 25/x
u_ash = 25
y_ash = u_ash / x_A

# Misty U=25 -> (10-x)(10-y) = 25 -> y = 10 - 25/(10-x)
u_misty = 25
y_misty = 10 - 25 / (10 - x_A)

plt.figure(figsize=(6, 6))

# Plot Contract Curve
plt.plot(x_A, y_contract, 'k-', linewidth=3, label='Contract Curve')

# Plot Indifference Curves
plt.plot(x_A, y_ash, 'b-', label="Ash's IC")
plt.plot(x_A, y_misty, 'r-', label="Misty's IC")

# Plot Endowment Point (Off-curve)
plt.plot(2, 8, 'ko')
plt.text(2.2, 8, 'Initial Endowment (Inefficient)', fontsize=10)

# Plot Equilibrium
plt.plot(5, 5, 'go')
plt.text(5.2, 4.5, 'Pareto Efficient', fontsize=10)

# Formatting
plt.title('The Edgeworth Box')
plt.xlabel("Ash's Burgers")
plt.ylabel("Ash's Fries")
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid(True, alpha=0.2)
plt.legend()

# Secondary Axis labels (Misty)
plt.text(9, 9.5, "Misty's Origin ->", color='red', ha='right')
plt.show()
```

**Graph Breakdown:**
*   **The Lens:** The space between the Blue curve (Ash) and Red curve (Misty) starting from the Endowment point. Any point inside the "Lens" is better for *both* parties.
*   **Tangency:** At the center (5,5), the curves kiss. They have the same slope. No more mutually beneficial trades are possible.

#### **(4) Key Takeaways**
*   **Efficiency $\neq$ Fairness:** One person having 100% of the food is "Pareto Efficient" (because to give some to the poor guy, you must hurt the rich guy). Economics focuses on efficiency; politics focuses on equity.
*   **Walras' Law:** If $N-1$ markets are in equilibrium, the last market *must* also be in equilibrium. (If the Burger market balances, the money market balances too).

#### **(5) Worked Example (Finding the Price)**
**Scenario:**
*   Ash: Endowment (10 Burgers, 0 Fries). Utility $U = B \cdot F$.
*   Misty: Endowment (0 Burgers, 10 Fries). Utility $U = B \cdot F$.
*   We need to find the relative price $P$ (Price of Burgers in terms of Fries). Let $P_F = 1$. Find $P_B = p$.

**Step 1: Calculate Ash's Wealth.**
*   Wealth $M_A = 10p + 0(1) = 10p$.

**Step 2: Calculate Ash's Demand.**
*   Cobb-Douglas: Spend 50% on Burgers.
*   $Demand_B^{Ash} = \frac{0.5 \times 10p}{p} = 5$.

**Step 3: Calculate Misty's Wealth.**
*   Wealth $M_M = 0p + 10(1) = 10$.

**Step 4: Calculate Misty's Demand.**
*   $Demand_B^{Misty} = \frac{0.5 \times 10}{p} = \frac{5}{p}$.

**Step 5: Market Clearing (Supply = Demand).**
*   Total Supply of Burgers = 10.
*   Total Demand = $5 + \frac{5}{p}$.
*   $10 = 5 + \frac{5}{p} \Rightarrow 5 = \frac{5}{p} \Rightarrow p = 1$.

**Result:** The price ratio is 1:1. Ash trades 5 Burgers for 5 Fries. They both end up with (5, 5).

### **Chapter 11: Market Failures (Externalities & Public Goods)**

#### **(1) Intuition First**
The invisible hand usually works... until it doesn't.

**Scenario 1 (Pollution):** Our truck emits smoke. It makes the burgers cheaper to cook, but Ash coughs while waiting in line.
*   **The Problem:** The truck owner (You) only cares about the cost of beef and gas (**Private Cost**). You ignore Ash's coughing (**Social Cost**).
*   **Result:** You produce *too many* smoke-heavy burgers. The market price is "too low" because it doesn't include the damage.

**Scenario 2 (The Park):** The city wants to build a park with benches for eating burgers.
*   **The Problem:** If Ash pays for the bench, Misty can sit on it for free.
*   **Public Good:** It is **Non-Rival** (Ash sitting doesn't stop Misty sitting) and **Non-Excludable** (You can't stop Misty from entering).
*   **Result:** No one builds the bench. Everyone waits for someone else to do it (**Free Riding**).

#### **(2) Formal Definition (Translated)**

*   **Negative Externality:** Social Marginal Cost ($SMC$) > Private Marginal Cost ($PMC$).
    *   *Solution:* **Pigouvian Tax**. Tax the firm exactly the amount of the external damage ($t = MD$). This forces them to "internalize" the damage.

*   **Public Good Provision (Samuelson Condition):**
    For private goods, we sum quantities horizontally ($Q_{total} = q_A + q_B$).
    For public goods, we sum **valuations** vertically.
    $$\sum MRS = MRT$$
    *   *Translation:* Since everyone consumes the *same* park, we add up how much Ash loves it + how much Misty loves it. If Total Love > Cost to Build, we build it.

#### **(3) Graphs & Relationships**

Visualizing the **Negative Externality (Deadweight Loss)**.

```python
import matplotlib.pyplot as plt
import numpy as np

Q = np.linspace(0, 10, 100)

# Demand (Social Benefit)
P_demand = 10 - Q

# Private Marginal Cost (Supply)
PMC = 2 + 0.5 * Q

# External Cost (Damage) = 2 per unit
# Social Marginal Cost = PMC + 2
SMC = 4 + 0.5 * Q

plt.figure(figsize=(8, 6))

# Plot Curves
plt.plot(Q, P_demand, label='Demand (Private Benefit)', color='blue')
plt.plot(Q, PMC, label='Private Supply (PMC)', color='green')
plt.plot(Q, SMC, label='Social Cost (SMC)', color='red')

# Market Equilibrium (PMC = Demand)
# 2 + 0.5Q = 10 - Q -> 1.5Q = 8 -> Q = 5.33
Q_mkt = 5.33
P_mkt = 4.66
plt.plot(Q_mkt, P_mkt, 'ko', label='Market Eq')

# Social Optimum (SMC = Demand)
# 4 + 0.5Q = 10 - Q -> 1.5Q = 6 -> Q = 4
Q_soc = 4
P_soc = 6
plt.plot(Q_soc, P_soc, 'go', label='Social Optimum')

# Deadweight Loss Shade
# Area between SMC and Demand, from Q_soc to Q_mkt
plt.fill_between(Q, SMC, P_demand, where=(Q >= Q_soc) & (Q <= Q_mkt), 
                 color='gray', alpha=0.3, label='Deadweight Loss')

plt.title('Negative Externality: Pollution')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.legend()
plt.show()
```

**Graph Breakdown:**
*   **Green Line (Supply):** What the firm pays.
*   **Red Line (Social Cost):** What society pays (Green + Pollution).
*   **Market Eq:** The firm ignores pollution, produces 5.33 units.

#### **(4) Worked Example (Public Good)**

Sum the marginal benefits (Vertical Sum).
$$MSB = (40 - Q) + (80 - Q) = 120 - 2Q$$

**Step 2: Equate to Marginal Cost.**
$$120 - 2Q = 100$$
$$2Q = 20 \Rightarrow Q^* = 10$$

# Study Guide

### **Part 1: The Consumer (Ash)**
*The Logic of "Wanting" vs. "Affording"*

1.  **The Budget Constraint:** Ash wants everything, but his wallet is finite. The **Budget Line** is the fence he cannot cross. Its slope ($-P_x/P_y$) tells him the market's trade-off (Opportunity Cost).
2.  **Preferences:** Ash knows what he likes.
    *   **Cobb-Douglas:** He likes variety (balance).
    *   **Perfect Substitutes:** He just wants the "kick" (Caffeine is Caffeine).
    *   **Perfect Complements:** He needs pairs (Burgers and Buns).
3.  **Optimization:** Ash finds the "Highest Mountain" (Indifference Curve) that touches his "Fence" (Budget Line).
    *   **The Golden Rule:** $MRS = \frac{P_x}{P_y}$. (Internal desire equals Market reality).
4.  **Demand:** When prices change, Ash reacts in two ways:
    *   **Substitution Effect:** "Burgers are cheaper, so I swap Fries for Burgers." (Always negative).
    *   **Income Effect:** "Burgers are cheaper, so I feel richer." (Buy more if Normal, buy less if Inferior).

### **Part 2: The Producer (The Truck)**
*The Logic of "Making" and "Selling"*

1.  **Technology:** The recipe. How we turn Labor ($L$) and Capital ($K$) into Burgers ($Q$).
    *   **Short Run:** Capital is fixed (One truck). Adding cooks leads to **Diminishing Marginal Returns**.
    *   **Long Run:** We can buy more trucks. **Returns to Scale** tells us if getting bigger makes us more efficient.
2.  **Cost Minimization:** We want the cheapest way to make 100 burgers.
    *   **Rule:** $\frac{MP_L}{MP_K} = \frac{w}{r}$. (Bang-per-buck of Labor = Bang-per-buck of Capital).
3.  **Profit Maximization:**
    *   **Marginal Cost (MC):** The cost of the *next* burger.
    *   **Supply Rule:** Keep cooking until Price ($P$) = Marginal Cost ($MC$).

### **Part 3: Market Structures (The Street)**
*The Logic of Competition*

1.  **Perfect Competition:** Many trucks. We are **Price Takers**. $P = MC$. Zero long-run profit.
2.  **Monopoly:** We are the *only* truck. We are **Price Makers**.
    *   **The Trap:** To sell more, we must lower the price on *everyone*. So, $MR < P$.
    *   **Rule:** Produce where $MR = MC$. Charge a markup based on elasticity.
    *   **Cost:** Deadweight Loss (We starve low-value customers to keep prices high).
3.  **Oligopoly:** Rivalry with "Burger Bros."
    *   **Cournot:** We fight over **Quantity**. (Strategic Substitutes).
    *   **Bertrand:** We fight over **Price**. (Strategic Complements $\to$ Zero Profit).
    *   **Stackelberg:** One moves first (Leader), the other reacts (Follower).

### **Part 4: The Big Picture (The City)**
*The Logic of Efficiency and Failure*

1.  **General Equilibrium:** Ash trades with Misty. They trade until their MRSs are equal. The result is **Pareto Efficient** (You can't help one without hurting the other).
2.  **Externalities:** Our smoke makes Ash cough.
    *   **Problem:** We ignore the Social Cost. We produce too much.
    *   **Solution:** Pigouvian Tax (Make us pay for the smoke).
3.  **Public Goods:** The City Park.
    *   **Problem:** Non-Rival + Non-Excludable. Everyone waits for someone else to build it (Free Riding).
    *   **Solution:** Sum the vertical willingness to pay ($\sum MRS = MRT$).
