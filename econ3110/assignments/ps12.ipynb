## Question 1 (6 points)

**State whether each of the following claims is true, false, or uncertain. If true, prove it. If false, provide a counterexample. If uncertain, explain what is at issue.**

**(a) (2 points) For any fixed endowment, all Pareto-efficient points are competitive equilibria.**

**False.** 
While the First Fundamental Theorem of Welfare Economics states that any competitive equilibrium (CE) is Pareto efficient (PE), the reverse is not generally true for a *fixed* endowment. The Second Fundamental Theorem states that any Pareto-efficient allocation can be supported as a competitive equilibrium *if* appropriate lump-sum transfers of wealth are made first. Without transfers, a specific endowment usually leads to a specific Competitive Equilibrium. Other points on the contract curve (which are Pareto efficient) are not equilibria for that specific initial endowment.

**(b) (2 points) With Cobb-Douglas preferences for both consumers, the contract curve is a straight line.**

**False.**
The contract curve is defined by the condition $MRS_A = MRS_B$. Let utility functions be $U_A = x_A^a y_A^b$ and $U_B = x_B^c y_B^d$. 
$$ MRS_A = \frac{a y_A}{b x_A}, \quad MRS_B = \frac{c y_B}{d x_B} $$
Equating them and substituting the feasibility constraints ($x_B = \bar{X} - x_A, y_B = \bar{Y} - y_A$):
$$ \frac{a y_A}{b x_A} = \frac{c (\bar{Y} - y_A)}{d (\bar{X} - x_A)} $$
This equation simplifies to a relationship between $y_A$ and $x_A$. Unless the preferences are identical (specifically if ratios $a/b = c/d$), this relationship describes a curve, not a straight line.

**(c) (2 points) The origins of the Edgeworth box are Pareto efficient.**

**True.**
At an origin (e.g., $O_A$), Agent A has zero goods and Agent B holds the entire endowment. To improve Agent A's utility, we must transfer goods from B to A. Assuming weakly monotone utility functions, taking goods away from Agent B will either lower B's utility or keep it constant (it cannot increase it). If we assume standard non-satiation (strictly monotone), B is made strictly worse off. Therefore, there is no way to make A better off without making B worse off, satisfying the definition of Pareto efficiency.

## Question 2 (5 points)

**Setup:**
- Production functions: $Q_B = \sqrt{C_B D_B}$ and $Q_I = \sqrt{C_I D_I}$.
- Endowment: $C = 150, D = 200$.
- Find the production contract curve $D_B(C_B)$ and derive the PPF.

**Solution:**
Efficiency in production requires the marginal rate of technical substitution (MRTS) to be equal across sectors:
$$ MRTS_B = \frac{MP_{C_B}}{MP_{D_B}} = \frac{0.5 C_B^{-0.5}D_B^{0.5}}{0.5 C_B^{0.5}D_B^{-0.5}} = \frac{D_B}{C_B} $$
$$ MRTS_I = \frac{D_I}{C_I} = \frac{200 - D_B}{150 - C_B} $$

Setting $MRTS_B = MRTS_I$:
$$ \frac{D_B}{C_B} = \frac{200 - D_B}{150 - C_B} $$
$$ D_B(150 - C_B) = C_B(200 - D_B) $$
$$ 150 D_B - C_B D_B = 200 C_B - C_B D_B $$
$$ 150 D_B = 200 C_B \implies D_B = \frac{4}{3} C_B $$

The **contract curve** is the line $D_B = \frac{4}{3} C_B$.

To derive the **PPF**, substitute the contract curve condition back into the production functions:
1. For the chatbot ($Q_B$):
$$ Q_B = \sqrt{C_B \left(\frac{4}{3}C_B\right)} = C_B \sqrt{\frac{4}{3}} $$
$$ \implies C_B = Q_B \sqrt{\frac{3}{4}} $$

2. For images ($Q_I$):
$$ Q_I = \sqrt{C_I D_I} = \sqrt{(150-C_B)(200 - \frac{4}{3}C_B)} $$
$$ Q_I = \sqrt{(150-C_B) \cdot \frac{4}{3}(150-C_B)} = (150-C_B)\sqrt{\frac{4}{3}} $$

Summing the output equations:
$$ Q_B + Q_I = C_B\sqrt{\frac{4}{3}} + (150-C_B)\sqrt{\frac{4}{3}} = 150\sqrt{\frac{4}{3}} $$
$$ Q_B + Q_I = 150 \cdot \frac{2}{\sqrt{3}} = 100\sqrt{3} \approx 173.2 $$

The PPF is linear: **$Q_B + Q_I = 100\sqrt{3}$**.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define PPF parameters
Q_total = 100 * np.sqrt(3)
Q_I = np.linspace(0, Q_total, 100)
Q_B = Q_total - Q_I

plt.figure(figsize=(6, 6))
plt.plot(Q_I, Q_B, label=r'PPF: $Q_B + Q_I = 100\sqrt{3}$', color='blue')
plt.title('Production Possibility Frontier')
plt.xlabel('$Q_I$ (Image Generation)')
plt.ylabel('$Q_B$ (Chatbot)')
plt.xlim(0, 180)
plt.ylim(0, 180)
plt.grid(True)
plt.legend()
plt.show()
```

## Question 3 (11 points)

**Setup:**
- $n=100$ trainers.
- Cost $C(q_i) = 5q_i$. Total Quality $Q = \sum q_i$.
- Utility $U_i = \frac{5}{2} \ln(Q) + w - 5q_i$.

### (a) (2 points) Socially optimal quality ($Q^{**}$)
Maximize the sum of utilities:
$$ W = \sum_{i=1}^n U_i = n \left[ \frac{5}{2} \ln(Q) + w \right] - 5Q $$
F.O.C. with respect to $Q$:
$$ \frac{d W}{d Q} = \frac{5n}{2Q} - 5 = 0 $$
$$ \frac{5(100)}{2Q} = 5 \implies \frac{250}{Q} = 5 \implies Q^{**} = 50 $$

### (b) (2 points) Symmetric Nash equilibrium ($Q^*$ and $q_i^*$)
Each trainer maximizes their own utility taking others' contributions as given:
$$ \max_{q_i} \left[ \frac{5}{2} \ln(q_i + Q_{-i}) + w - 5q_i \right] $$
F.O.C.:
$$ \frac{5}{2(q_i + Q_{-i})} - 5 = 0 \implies \frac{5}{2Q} = 5 $$
$$ Q^* = \frac{1}{2} = 0.5 $$
Each trainer contributes $q_i = 0.5/100 = 0.005$.

### (c) (2 points) Compare $Q^*$ with $Q^{**}$. Intuition?
- $Q^* = 0.5$
- $Q^{**} = 50$
<br>
The socially optimal level is 100 times higher than the Nash equilibrium level. This is a classic **public good problem (free-riding)**. Individuals only consider their private marginal benefit (which is $1/n$ of the social marginal benefit) when choosing effort, leading to severe under-provision of the public good.

### (d) (3 points) Cost rises to $C(q_i) = 25q_i$.
- **Social Optimum:** Max $\frac{5n}{2} \ln Q - 25Q \implies \frac{250}{Q} = 25 \implies Q^{**} = 10$.
- **Nash Equilibrium:** $\frac{5}{2Q} = 25 \implies Q^* = \frac{5}{50} = 0.1$.
- **Comparison:** 
  - Previous $Q^* = 0.5$, New $Q^* = 0.1$. 
  - The cost increased by a factor of 5 ($5 \to 25$). The quality decreased by a factor of 5 ($0.5 \to 0.1$).
  - **Yes**, the model quality falls proportionally.

### (e) (2 points) Asymmetric case ($n=2$). Cost $C(q)=5q$.
Utilities: 
$ U_A = 2 \ln(Q) + w_A - 5q_A $
$ U_B = 3 \ln(Q) + w_B - 5q_B $
<br>
FOCs (assuming interior solutions):
A: $ \frac{2}{Q} = 5 \implies Q = 0.4 $
B: $ \frac{3}{Q} = 5 \implies Q = 0.6 $
<br>
These target Q levels are inconsistent. In public goods games with linear costs, the individual with the highest valuation provides the good until their marginal benefit equals marginal cost. The lower-valuation individual free-rides completely.
- Since B wants $Q=0.6$ and A only wants $Q=0.4$, B will contribute enough to reach 0.6. At $Q=0.6$, A's marginal benefit is $2/0.6 = 3.33 < 5$, so A contributes 0.
- **Result:** $\tilde{Q} = 0.6$ (with $q_A=0, q_B=0.6$).
- **Comparison:** In part (b) with $n=2$, the parameter was 2.5 (avg of 2 and 3). The symmetric result would be $2.5/5 = 0.5$. Here $\tilde{Q}=0.6$. The quality is **higher**. This is because the aggregate provision is determined by the maximum valuation in the group, which exceeds the average valuation.

## Question 4 (8 points)

**Payoffs:**
- **Dash (Ken) vs Hadouken (Ryu):** $(-1, 1)$ [Correct Counter]
- **Dash (Ken) vs Shoryuken (Ryu):** $(3, -3)$ [Incorrect]
- **Jump (Ken) vs Shoryuken (Ryu):** $(-2, 2)$ [Correct Counter]
- **Jump (Ken) vs Hadouken (Ryu):** $(3, -3)$ [Incorrect]

**(a) (3 points) Subgame Perfect Nash Equilibrium (SPNE)**
We use backward induction:
1. **Ryu's nodes:**
   - If Ken **Jumps**: Ryu chooses between Hadouken (-3) and Shoryuken (2). Ryu chooses **Shoryuken**.
   - If Ken **Dashes**: Ryu chooses between Hadouken (1) and Shoryuken (-3). Ryu chooses **Hadouken**.
2. **Ken's node:**
   - Ken anticipates Ryu's responses: 
     - If Ken Jumps $\to$ Ryu Shoryuken $\to$ Ken gets -2.
     - If Ken Dashes $\to$ Ryu Hadouken $\to$ Ken gets -1.
   - Ken compares -2 and -1. Ken chooses **Dash**.

**SPNE:** Ken Dashes; Ryu plays (Hadouken if Dash, Shoryuken if Jump).

**(b) (2 points) Simultaneous Move: Pure Strategy Nash Equilibrium**
Payoff Matrix (Ken, Ryu):

| Ken \ Ryu | Hadouken | Shoryuken |
|---|---|---|
| **Dash** | $(-1, 1)$ | $(3, -3)$ |
| **Jump** | $(3, -3)$ | $(-2, 2)$ |

Checking for Pure Strategy NE:
- If Ken Dash: Ryu picks Hadouken ($1 > -3$).
- If Ryu Hadouken: Ken picks Jump ($3 > -1$).
- If Ken Jump: Ryu picks Shoryuken ($2 > -3$).
- If Ryu Shoryuken: Ken picks Dash ($3 > -2$).

There is a cycle. **No, there is no Nash equilibrium in pure strategies.**

**(c) (3 points) Simultaneous Move: Mixed Strategy Nash Equilibrium**
Let $p$ be the probability Ken Dashes, and $q$ be the probability Ryu uses Hadouken.

1. **Indifference for Ken:**
   $$ E[U_{Dash}] = E[U_{Jump}] $$
   $$ -1(q) + 3(1-q) = 3(q) - 2(1-q) $$
   $$ 3 - 4q = 5q - 2 $$
   $$ 5 = 9q \implies q = \frac{5}{9} $$

2. **Indifference for Ryu:**
   $$ E[U_{Hadouken}] = E[U_{Shoryuken}] $$
   $$ 1(p) - 3(1-p) = -3(p) + 2(1-p) $$
   $$ 4p - 3 = 2 - 5p $$
   $$ 9p = 5 \implies p = \frac{5}{9} $$

**MSNE:** Ken Dashes with probability $5/9$ (Jumps with $4/9$), and Ryu uses Hadouken with probability $5/9$ (Shoryuken with $4/9$).
