## Question 1 (8 points)

**(a) Nash Equilibrium:**  
A set of strategies (one for each player) is a Nash Equilibrium if no player can unilaterally change their strategy to increase their payoff, holding the strategies of all other players constant. Formally, a strategy profile $s^*$ is a Nash Equilibrium if $u_i(s_i^*, s_{-i}^*) \geq u_i(s_i, s_{-i}^*)$ for all strategies $s_i$ and all players $i$.

**(b) A Subgame:**  
In an extensive-form game, a subgame is a subset of the game that starts at a specific node (a singleton information set), includes all subsequent nodes and decision branches derived from that node, and preserves the information structure of the original game.

**(c) Subgame-Perfect Nash Equilibrium (SPNE):**  
A strategy profile is a Subgame-Perfect Nash Equilibrium if it induces a Nash Equilibrium in every possible subgame of the original game. This concept eliminates non-credible threats.

**(d) Stackelberg Competition:**  
A model of imperfect competition where firms compete on quantity (like Cournot), but moves are sequential. There is a "leader" firm that moves first, and a "follower" firm that moves sequentially after observing the leader's output.

**(e) Bertrand Competition:**  
A model of imperfect competition where firms compete by setting prices simultaneously. In the standard model with homogeneous goods, this leads to price equaling marginal cost ($P=MC$).

## Question 3 (6 points)

**Parameters:** $a > 0$, $b > c > 0$.  
**Inverse Demand:**  
$p_1 = a - bq_1 - cq_2$  
$p_2 = a - bq_2 - cq_1$  
**Costs:** $MC=0, FC=0$. 

Cournot Output ($q$): $$\frac{a}{2b + c}$$ <br>
Cournot Price ($p$): $$\frac{ab}{2b + c}$$ <br>
Bertrand Price ($p$): $$\frac{a(b - c)}{2b - c}$$ <br>
Bertrand Output ($q$): $$\frac{ab}{2b^2 + bc - c^2}$$

### Answers for Q3

**(a) Cournot Equilibrium**  
From the derivation above:
$$ q_1^C = q_2^C = \frac{a}{2b + c} $$
$$ p_1^C = p_2^C = \frac{ab}{2b + c} $$

**(b) Bertrand Equilibrium**  
Using the direct demand functions derived from inverting the system:
$$ p_1^B = p_2^B = \frac{a(b-c)}{2b - c} $$
$$ q_1^B = q_2^B = \frac{a b}{(b+c)(2b-c)} $$

**(c) Comparison and Intuition**  
- **Output:** Cournot quantity is lower than Bertrand quantity.  
- **Price:** Cournot price is higher than Bertrand price.  
- **Intuition:** In Cournot competition, firms choose quantities, taking the other's quantity as given. This allows them to maintain higher prices by restricting output. In Bertrand competition, firms compete on price. Even with imperfect substitutes ($b > c$), the incentive to undercut the rival to steal market share is stronger, driving prices down closer to marginal cost (though not exactly to MC=0 due to product differentiation).

## Question 4 (6 points)

**Utility Functions:**
- $u_1(x_A, x_B) = \sqrt{x_A} + \sqrt{x_B}$
- $u_2(x_A, x_B) = x_A + x_B$

**Endowments:** $e_1 = (1, 0), e_2 = (0, 1)$. Total $A=1, B=1$.

**(a) Set of Pareto-efficient allocations**  
Pareto efficiency requires $MRS_1 = MRS_2$.
- $MRS_1 = \frac{MU_A}{MU_B} = \frac{0.5/\sqrt{x_{1A}}}{0.5/\sqrt{x_{1B}}} = \sqrt{\frac{x_{1B}}{x_{1A}}}$
- $MRS_2 = 1$ (Perfect substitutes)

Setting $MRS_1 = MRS_2$:
$$ \sqrt{\frac{x_{1B}}{x_{1A}}} = 1 \implies x_{1B} = x_{1A} $$
The Pareto set is the diagonal line where Farmer 1 consumes equal amounts of Apples and Bananas ($x_{1A} = x_{1B}$). Consequently, Farmer 2 also consumes equal amounts ($x_{2A} = x_{2B}$).

**(b) Edgeworth Box Visualization**

```python
import matplotlib.pyplot as plt
import numpy as np

# Box dimensions
total_A = 1.0
total_B = 1.0

# Grid for plotting
x = np.linspace(0.01, 0.99, 100)
y = np.linspace(0.01, 0.99, 100)
X, Y = np.meshgrid(x, y)

# Utility functions
# u1 depends on x1_A (X) and x1_B (Y)
U1 = np.sqrt(X) + np.sqrt(Y)

# u2 depends on x2_A (1-X) and x2_B (1-Y)
U2 = (1 - X) + (1 - Y)

fig, ax = plt.subplots(figsize=(6, 6))

# Plot Indifference Curves
cp1 = ax.contour(X, Y, U1, levels=15, cmap='Blues', alpha=0.6)
cp2 = ax.contour(X, Y, U2, levels=15, cmap='Reds', alpha=0.6)

# Plot Contract Curve (Diagonal)
ax.plot([0, 1], [0, 1], 'k--', linewidth=2, label='Pareto Set (x_1A = x_1B)')

# Endowment Point
ax.plot(1, 0, 'go', markersize=10, label='Endowment (e1)')

# Labels
ax.set_xlabel("Farmer 1 Apples ($x_{1A}$)")
ax.set_ylabel("Farmer 1 Bananas ($x_{1B}$)")
ax.set_title("Q4(b): Edgeworth Box")
ax.legend(loc='lower right')

plt.grid(True)
plt.show()
```

**(c) Analysis with $u_2(x_A, x_B) = x_A^{1+\epsilon} + x_B^{1+\epsilon}$**  
If $\epsilon > 0$, Farmer 2's utility function becomes strictly convex. This implies their indifference curves (relative to the origin $O_2$) are **bowed out** (concave to the origin $O_2$, or simply "bowed towards the middle of the box"). Farmer 2 prefers extreme bundles (all Apples or all Bananas) over mixtures.

On the diagram, Farmer 2's indifference curves would look like quarter-circles centered away from $O_2$. The tangency condition with Farmer 1's convex preferences (standard convex-to-origin ICs) would likely occur at minimums rather than maximums, or trade would be pushed to the **boundaries**.

**Is the Pareto set unchanged?**  
**No.** Because Farmer 2 prefers extremes, efficiency requires moving to the edges of the box. The Pareto efficient allocations will likely consist of the edges where Farmer 2 consumes only one good (e.g., the bottom and right edges of the box from Farmer 1's perspective).

## Question 5 (6 points)

**Preferences:**
- Kanto ($K$): $u_K = D_K + 2A_K$ ($MRS_{DA}^K = 1/2$)
- Johto ($J$): $u_J = 2D_J + A_J$ ($MRS_{DA}^J = 2$)

Note: $MRS_{DA} = MU_D / MU_A$. The marginal rate of substitution is constant for both agents (linear indifference curves).
- K values A relatively more ($1 D$ trades for $0.5 A$ to keep utility constant).
- J values D relatively more ($1 D$ trades for $2 A$ to keep utility constant).

Since $MRS^K < MRS^J$ everywhere ($0.5 < 2$), mutually beneficial trade exists until one agent hits a constraint (runs out of a good). Efficiency requires the agent who values a good most to hold it.
- K should hold Apples.
- J should hold Dratinis.

**(a) Pareto-efficient allocations and Contract Curve (Initial Endowment)**
Total Endowments: $D_{tot}=1000, A_{tot}=1000$.
Pareto Set consists of allocations where K holds as many Abras as possible and J holds as many Dratinis as possible.
1. **Top Edge:** K has all 1000 Abras ($A_K=1000$), J has 0. ($D_K$ varies).
2. **Left Edge:** J has all 1000 Dratinis ($D_K=0$), K has 0. ($A_K$ varies).

The **Contract Curve** is the subset of Pareto efficient points reachable via mutually beneficial trade from the endowment $(D_K=1000, A_K=0)$.
- At endowment: $u_K = 1000$, $u_J = 1000$.
- Trade Logic: K trades D for A. J trades A for D.
- Resulting Equilibrium: K ends up with all Abras (and some Dratinis if A is cheaper than relative value) or J ends up with all Dratinis. Given the symmetry and linear preferences, they likely move to the corner $(D_K=0, A_K=1000)$.

```python
# Visualizing Q5 (a)
fig, ax = plt.subplots(figsize=(6, 6))

# Dimensions 1000x1000
ax.set_xlim(0, 1000)
ax.set_ylim(0, 1000)

# Pareto Set (L-shape)
# Top Edge: K has 1000 Abras (y=1000)
ax.plot([0, 1000], [1000, 1000], 'r-', linewidth=4, label='Pareto Set')
# Left Edge: K has 0 Dratinis (x=0)
ax.plot([0, 0], [0, 1000], 'r-', linewidth=4)

# Endowment Q5a (1000 Dratinis for K)
ax.plot(1000, 0, 'go', markersize=10, label='Endowment (a)')

ax.set_xlabel("Kanto Dratinis ($D_K$)")
ax.set_ylabel("Kanto Abras ($A_K$)")
ax.set_title("Q5(a): Edgeworth Box (1000x1000)")
ax.legend()
plt.grid(True)
plt.show()
```

**(b) Set of allocations for Contract Curve**
The contract curve connects the endowment to the Pareto set. However, technically the Contract Curve is the set of Pareto Optimal points where $u_i \geq u_{i,0}$.
From $e=(1000, 0)$, they trade until K has all Abras. The efficient point $(0, 1000)$ gives $u_K=2000, u_J=2000$. Both are better off.
