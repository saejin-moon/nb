### Question 1 (5 points)

#### (a) $F(K, L) = K + L$
The inputs are perfect substitutes. The firm will use only the cheaper input. <br>
If $w < r$, then $(K^*, L^*) = (0, Q)$. <br>
If $w > r$, then $(K^*, L^*) = (Q, 0)$. <br>
If $w = r$, any combination of $K$ and $L$ such that $K + L = Q$ is cost-minimizing. <br>

#### (b) $F(K, L) = min{K, L}$
The inputs are perfect complements. To produce $Q$ units, the firm needs $K = Q$ and $L = Q$. <br>
So, $(K^*, L^*) = (Q, Q)$.

#### (c) $F(K, L) = K^\alpha L^(1-\alpha)$ <br>
The marginal rate of technical substitution is $MRTS = \frac{\alpha}{(1-\alpha)} * \frac{L}{K}$. <br> 
Setting $MRTS = \frac{w}{r}$, we get $L = K * \frac{(1-\alpha)}{\alpha} * \frac{r}{w}$. <br>
Substituting this into the production function: $Q = K^\alpha * (K * \frac{1-\alpha}{\alpha} * (\frac{r}{w}))^(1-\alpha)$ <br>
Solving for $K^*$ and then $L^*$: <br>
$K^* = Q \left(\frac{\alpha r}{(1-\alpha)w}\right)^{\alpha-1}$ <br>
$L^* = Q \left(\frac{\alpha w}{(1-\alpha)r}\right)^{-\alpha}$ <br>

#### (d) $F(K, L) = ln(K) + ln(L)$
This is a Cobb-Douglas like function. The $MRTS = \frac{K}{L}$. <br>
Setting $MRTS = \frac{w}{r}$ gives $\frac{K}{L} = \frac{w}{r}$, so $K = L(\frac{w}{r})$. <br>
Substituting into the production function: $Q = ln(L(\frac{w}{r})) + ln(L) = ln(L^2 * \frac{w}{r})$. <br>
$e^Q = L^2 * (\frac{w}{r})$. <br>
$L^* = \sqrt{\frac{r e^Q}{w}}$ <br>
$K^* = \sqrt{\frac{w e^Q}{r}}$ <br>

#### (e) $F(K, L) = K + L^2$
The $MRTS = \frac{1}{2L}$. <br>
Setting $MRTS = \frac{w}{r}$ gives $\frac{1}{2L} = \frac{w}{r}$, so $L = \frac{r}{2w}$. <br>
If $Q > (\frac{r}{2w})^2$, then $L^* = \frac{r}{2w}$ and $K^* = Q - (\frac{r}{2w})^2$. <br>
If $Q <= (\frac{r}{2w})^2$, then it is a corner solution with $K=0$, so $L^* = \sqrt{Q}$ and $K^*=0$. <br>

### Question 3 (6 points)

#### (a)
Total Cost: $TC(q) = F + q + q^2$ <br>
Average Total Cost: $ATC(q) = \frac{F}{q} + 1 + q$

#### (b)
In long-run competitive equilibrium, firms produce at the minimum of ATC. <br>
To find the minimum ATC, we set its derivative with respect to $q$ to zero: <br>
$\frac{dATC}{dq} = \frac{-F}{q^2} + 1 = 0$ <br>
$q^* = \sqrt{F}$ <br>
The minimum ATC is $ATC(\sqrt{F}) = F/\sqrt{F} + 1 + \sqrt{F} = 2\sqrt{F} + 1$. <br>
The equilibrium price is $P^* = 2\sqrt{F} + 1$. <br>
At this price, total market quantity is $Q = \frac{a - (2\sqrt{F} + 1)}{b}$. <br>
The number of firms is $n^* = \frac{Q}{q^*} = \frac{a - 2\sqrt{F} - 1}{b\sqrt{F}}$. <br>

#### (c)
The cartel's total revenue is $TR(Q) = (a - bQ)Q$. <br>
Marginal revenue is $MR(Q) = a - 2bQ$. <br>
The market marginal cost is the horizontal sum of individual MCs. For $n$ firms, the total cost is $n(F+q+q^2)$. <br>
With $Q=nq$, the market total cost is $C(Q) = nF + Q + \frac{Q^2}{n}$. <br>
Market marginal cost is $MC(Q) = 1 + \frac{2Q}{n}$. <br>
Setting $MR = MC$: <br>
$a - 2bQ = 1 + \frac{2Q}{n}$ <br>
$Q^c = \frac{a-1}{2b + 2/n}$ <br>
Per-firm quantity is $q^c = \frac{Q^c}{n} = \frac{a-1}{2bn + 2}$. <br>

#### (d)

```python
a, b, F = 10, 1, 1
q_star = np.sqrt(F)
p_star = 2*np.sqrt(F) + 1
Q_comp = (a - p_star)/b
n_star = Q_comp/q_star

Q = np.linspace(0, a/b, 100)
demand = a - b*Q
mc_comp = 1 + 2*Q/n_star
mc_cartel = 1 + 2*Q/n_star
mr = a - 2*b*Q
Q_cartel = (a-1)/(2*b + 2/n_star)
p_cartel = a - b*Q_cartel

plt.figure(figsize=(8, 6))
plt.plot(Q, demand, label='Demand')
plt.plot(Q, mr, label='Marginal Revenue')
plt.plot(Q, mc_comp, label='Competitive MC')
plt.scatter(Q_comp, p_star, color='red', label=f'Competitive Eq: Q={Q_comp:.2f}, P={p_star:.2f}')
plt.scatter(Q_cartel, p_cartel, color='green', label=f'Cartel Eq: Q={Q_cartel:.2f}, P={p_cartel:.2f}')

# Deadweight Loss
q_fill = np.linspace(Q_cartel, Q_comp, 100)
p_demand_fill = a - b * q_fill
p_mc_fill = 1 + 2*q_fill/n_star
plt.fill_between(q_fill, p_demand_fill, p_mc_fill, color='gray', alpha=0.5, label='Deadweight Loss')

plt.xlabel('Quantity (Q)')
plt.ylabel('Price (P)')
plt.title('Deadweight Loss of Cartel')
plt.legend()
plt.grid(True)
plt.show()
```

#### (e)
A cartel with free entry will have firms enter until the cartel's profit is zero. The cartel maximizes industry profit by setting MR = MC, resulting in a higher price and lower quantity than the competitive outcome. The competitive equilibrium drives price down to the minimum of AC, ensuring zero economic profit. The cartel price will be above this minimum AC, but entry will occur until the cartel's total revenue equals its total cost, which occurs at a higher price and lower quantity than the competitive equilibrium, hence lower consumer surplus.

### Question 4 (6 points)

#### (b)
Market demand is $Q(p) = \alpha q_h(p) + (1-\alpha)q_l(p) = \alpha(60-p) + (1-\alpha)(20-4p)$. <br>
$Q(p) = 60\alpha - \alpha p + 20 - 4p - 20\alpha + 4\alpha p = (40\alpha + 20) + (3\alpha - 4)p$

```python
p = np.linspace(0, 20, 100)
alphas = [0, 0.5, 1]

plt.figure(figsize=(8, 6))
for alpha in alphas:
    q = (40*alpha + 20) + (3*alpha - 4)*p
    plt.plot(q, p, label=f'alpha = {alpha}')

plt.xlabel('Quantity (Q)')
plt.ylabel('Price (p)')
plt.title('Market Demand for Magical Potions')
plt.legend()
plt.grid(True)
plt.show()
```

#### (c)
The mage sets one price. Total demand is $Q(p) = (40\alpha + 20) + (3\alpha - 4)p$.- <br>
The inverse demand is $p(Q) = \frac{Q - 40\alpha - 20}{3\alpha - 4}$. <br>
Total revenue is $TR(Q) = Q * \frac{Q - 40\alpha - 20}{3\alpha - 4}$. <br>
$MR = \frac{2Q - 40\alpha - 20}{3\alpha - 4}$. <br>
$MC = 4Q - 4$. <br>
Setting MR = MC and solving for Q will give the optimal quantity, and then the price. The solution depends on $\alpha$.

#### (d)
New cost function $C(q) = 8q$, so $MC = 8$. <br>
Setting MR = MC: <br>
$\frac{2Q - 40\alpha - 20}{3\alpha - 4} = 8$ <br>
$2Q - 40\alpha - 20 = 24\alpha - 32$ <br>
$2Q = 64\alpha - 12$ <br>
$Q^* = 32\alpha - 6$. <br>
$p^* = \frac{(32\alpha - 6) - 40\alpha - 20}{3\alpha - 4} = \frac{-8\alpha - 26}{3\alpha - 4}$. <br>
$q_h^* = 60 - p^* \quad \text{and} \quad q_l^* = 20 - 4p^*$

#### (e)
For type-h consumers: $MC_h = 12$. Inverse demand $p_h = 60 - q_h$. $MR_h = 60 - 2q_h$. <br>
$60 - 2q_h = 12 \implies 2q_h = 48 \implies q_h^* = 24$. <br>
$p_h^* = 60 - 24 = 36$. <br>

For type-l consumers: $MC_l = 6$. Inverse demand $p_l = 5 - 0.25q_l$. $MR_l = 5 - 0.5q_l$. <br>
$5 - 0.5q_l = 6 \implies -1 = 0.5q_l \implies q_l = -2$. This means the mage does not serve the low-value customers as MC is above their max willingness to pay (5). <br>
So, $q_l^* = 0$ and $p_l^*$ is not applicable. <br>

#### (f)
The price $P^*$ and quantities $q_h^*, q_l^*$ from part (d) are: <br>
$p^* = \frac{-8\alpha - 26}{3\alpha - 4}$ <br>
$q_h^* = 60 - p^*$ <br>
$q_l^* = 20 - 4p^*$ <br>

The change in CS for type-h is the area of the trapezoid between $P^*$ and $P_h^*=36$: <br>
$\Delta CS_h = 0.5 \times (q_h^* + 24) \times (P^* - 36)$ <br>

The change in CS for type-l is the negative of the CS in part (d) (since $CS_l$ in part (e) is 0): <br>
$CS_l(d) = 0.5 \times \max(0, q_l^*) \times (5 - P^*)$ <br>
$\Delta CS_l = -CS_l(d)$ <br>

```python
alpha_vals = np.linspace(0, 1, 100)
p_star = (-8*alpha_vals - 26)/(3*alpha_vals - 4)
q_h_star = 60 - p_star
q_l_star = 20 - 4*p_star

delta_cs_h = 0.5 * (q_h_star + 24) * (p_star - 36)

cs_l_d = 0.5 * np.maximum(0, q_l_star) * (5 - p_star)
delta_cs_l = -cs_l_d

delta_welfare = alpha_vals * delta_cs_h + (1 - alpha_vals) * delta_cs_l

plt.figure(figsize=(8, 6))
plt.plot(alpha_vals, delta_welfare, label='Change in Consumer Welfare')
plt.xlabel('Alpha ()')
plt.ylabel('Change in Welfare')
plt.title('Change in Consumer Welfare from (d) to (e)')
plt.legend()
plt.grid(True)
plt.show()
```

### Question 6 (4 points)

#### (a)
For a Cobb-Douglas production function $Q = K^{1/3}L^{1/3}$, the $MRTS = \frac{K}{L}$. <br>
To minimize cost for one unit of output, we set $MRTS = \frac{w}{r}$, so $frac{K}{L} = \frac{w}{r}$, or $K = L(\frac{w}{r})$. <br>
$1 = (L(\frac{w}{r}))^{1/3} L^{1/3} = L^{2/3} (\frac{w}{r})^{1/3}$. <br>
$L^{2/3} = (\frac{r}{w})^{1/3} \implies L = (\frac{r}{w})^{1/2}$. <br>
$K = (\frac{r}{w})^{1/2} (\frac{w}{r}) = (\frac{w}{r})^{1/2}$. <br>
This applies to both apples and bananas. So $(K_A^*, L_A^*) = ((\frac{w}{r})^{1/2}, (\frac{r}{w})^{1/2})$ and $(K_B^*, L_B^*) = ((\frac{w}{r})^{1/2}, (\frac{r}{w})^{1/2})$. <br>

#### (b)
The cost of producing one unit is $c(w,r) = wL + rK = w(\frac{r}{w})^{1/2} + r(\frac{w}{r})^{1/2} = (wr)^{1/2} + (wr)^{1/2} = 2\sqrt{wr}$. <br>
Profit maximization requires price equals unit cost. <br>
$p_A = 2\sqrt{wr}$ and $p_B = 2\sqrt{wr}$. <br>
This implies $p_A = p_B$. The factor prices are not uniquely determined by output prices in this case, but we have the relationship $wr = (p_A/2)^2 = (p_B/2)^2$.<br>

#### (c)
If $p_A$ increases, for the equality $p_A = 2\sqrt{wr}$ to hold, the product $wr$ must increase. This means either $w$, or $r$, or both must increase.
If only households that own capital do not supply labor, and the rest supply labor, an increase in $r$ benefits the capital owners, while an increase in $w$ benefits the labor suppliers. Without more information on how the change in $p_A$ is distributed between $w$ and $r$, we cannot definitively say who benefits more. However, it's likely that the return to the factor used more intensively in apple production will increase more. Since both industries have the same production function, we cannot make this distinction. But an increase in $p_A$ will lead to an expansion of the apple industry, increasing demand for both K and L, thus likely increasing both w and r. So both groups would benefit.
